## 1. Teor√≠a: Presentaci√≥n del Curso y Ciencia de Datos en el Deporte

### ¬øQu√© es la Ciencia de Datos?

La ciencia de datos es un campo interdisciplinario que utiliza m√©todos cient√≠ficos, procesos, algoritmos y sistemas para extraer conocimiento y percepciones de datos estructurados y no estructurados.

### Ciencia de Datos en el Deporte

En el contexto deportivo, la ciencia de datos nos permite:

1. **An√°lisis de Rendimiento**: Evaluar el desempe√±o de jugadores y equipos
2. **Predicci√≥n de Resultados**: Crear modelos que anticipen resultados de partidos
3. **Optimizaci√≥n T√°ctica**: Identificar patrones y estrategias exitosas
4. **Gesti√≥n de Recursos**: Optimizar fichajes y gesti√≥n de plantillas
5. **Experiencia del Aficionado**: Mejorar la experiencia mediante estad√≠sticas avanzadas

### El F√∫tbol como Fuente de Datos

El f√∫tbol genera una gran cantidad de datos:
- **Resultados de partidos**: Marcadores, fechas, equipos
- **Eventos del juego**: Goles, tarjetas, faltas, corners
- **Datos de jugadores**: Posiciones, pases, disparos
- **Datos temporales**: Temporadas, competiciones, fechas

## 3. Pr√°ctica: Exploraci√≥n del Dataset "champs"

### Carga del Dataset

Vamos a cargar el dataset "champs" que contiene informaci√≥n sobre la UEFA Champions League. Este dataset t√≠picamente contiene informaci√≥n sobre partidos, equipos, goles y resultados.

**Nota**: Para este ejemplo, crearemos un dataset simulado que represente la estructura t√≠pica del dataset "champs". En un entorno real, cargar√≠as el archivo CSV directamente.

````python
# Simulamos la carga del dataset "champs"
# En la pr√°ctica real, usar√≠as: df_champs = pd.read_csv('path/to/champs.csv')

# Creamos datos simulados que representen la estructura t√≠pica
np.random.seed(42)

# Generar datos de ejemplo
equipos_champions = [
    'Barcelona', 'Real Madrid', 'Bayern Munich', 'Manchester City',
    'Liverpool', 'Paris Saint-Germain', 'Chelsea', 'Juventus',
    'Manchester United', 'Arsenal', 'Atletico Madrid', 'Borussia Dortmund'
]

temporadas = ['2020-21', '2021-22', '2022-23']
fases = ['Group Stage', 'Round of 16', 'Quarter-finals', 'Semi-finals', 'Final']

# Generar datos simulados
n_partidos = 200
datos_champs = []

for i in range(n_partidos):
    equipo_local = np.random.choice(equipos_champions)
    equipo_visitante = np.random.choice([e for e in equipos_champions if e != equipo_local])
    
    goles_local = np.random.poisson(1.5)
    goles_visitante = np.random.poisson(1.2)
    
    datos_champs.append({
        'season': np.random.choice(temporadas),
        'stage': np.random.choice(fases),
        'date': f"2022-{np.random.randint(1,13):02d}-{np.random.randint(1,29):02d}",
        'home_team': equipo_local,
        'away_team': equipo_visitante,
        'home_goals': goles_local,
        'away_goals': goles_visitante,
        'home_possession': np.random.randint(35, 75),
        'away_possession': 100 - np.random.randint(35, 75)
    })

df_champs = pd.DataFrame(datos_champs)
print("‚úÖ Dataset 'champs' cargado exitosamente")
print(f"üìä Dimensiones: {df_champs.shape}")
````

### Visualizaciones B√°sicas

Creemos algunas visualizaciones para entender mejor los datos:

````python
# Crear visualizaciones con seaborn
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Gr√°fica 1: Distribuci√≥n de goles totales por partido
sns.histplot(data=df_champs, x='total_goals', bins=range(0, df_champs['total_goals'].max()+2), 
             ax=axes[0,0], kde=True, color='skyblue')
axes[0,0].set_title('Distribuci√≥n de Goles Totales por Partido', fontsize=14, fontweight='bold')
axes[0,0].set_xlabel('Goles Totales')
axes[0,0].set_ylabel('Frecuencia')

# Gr√°fica 2: Goles locales vs visitantes con l√≠nea de tendencia
sns.scatterplot(data=df_champs, x='home_goals', y='away_goals', ax=axes[0,1], 
                alpha=0.7, s=60)
sns.regplot(data=df_champs, x='home_goals', y='away_goals', ax=axes[0,1], 
            scatter=False, color='red', line_kws={'linestyle': '--'})
axes[0,1].set_title('Goles Locales vs Goles Visitantes', fontsize=14, fontweight='bold')
axes[0,1].set_xlabel('Goles Locales')
axes[0,1].set_ylabel('Goles Visitantes')

# Gr√°fica 3: Partidos por temporada
sns.countplot(data=df_champs, x='season', ax=axes[1,0], palette='viridis')
axes[1,0].set_title('Partidos por Temporada', fontsize=14, fontweight='bold')
axes[1,0].set_xlabel('Temporada')
axes[1,0].set_ylabel('N√∫mero de Partidos')
axes[1,0].tick_params(axis='x', rotation=45)

# Gr√°fica 4: Partidos por fase del torneo
sns.countplot(data=df_champs, x='stage', ax=axes[1,1], palette='Set2')
axes[1,1].set_title('Partidos por Fase del Torneo', fontsize=14, fontweight='bold')
axes[1,1].set_xlabel('Fase')
axes[1,1].set_ylabel('N√∫mero de Partidos')
axes[1,1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()
````

### An√°lisis Exploratorio Avanzado

Realicemos algunos an√°lisis m√°s profundos:

````python
# An√°lisis de ventaja de local con seaborn
partidos_local = len(df_champs[df_champs['home_goals'] > df_champs['away_goals']])
partidos_visitante = len(df_champs[df_champs['away_goals'] > df_champs['home_goals']])
partidos_empate = len(df_champs[df_champs['home_goals'] == df_champs['away_goals']])

print("üèüÔ∏è An√°lisis de Ventaja de Local:")
print(f"Victorias locales: {partidos_local} ({partidos_local/len(df_champs)*100:.1f}%)")
print(f"Victorias visitantes: {partidos_visitante} ({partidos_visitante/len(df_champs)*100:.1f}%)")
print(f"Empates: {partidos_empate} ({partidos_empate/len(df_champs)*100:.1f}%)")

# Crear DataFrame para visualizaci√≥n
resultados_data = pd.DataFrame({
    'Resultado': ['Victorias Locales', 'Empates', 'Victorias Visitantes'],
    'Cantidad': [partidos_local, partidos_empate, partidos_visitante]
})

# Visualizaci√≥n con seaborn
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

# Gr√°fico de barras
sns.barplot(data=resultados_data, x='Resultado', y='Cantidad', 
            palette=['#2E8B57', '#FFD700', '#DC143C'], ax=ax1)
ax1.set_title('Distribuci√≥n de Resultados', fontsize=14, fontweight='bold')
ax1.set_ylabel('N√∫mero de Partidos')
ax1.tick_params(axis='x', rotation=45)

# Gr√°fico de pastel con matplotlib (seaborn no tiene pie charts)
ax2.pie(resultados_data['Cantidad'], labels=resultados_data['Resultado'], 
        colors=['#2E8B57', '#FFD700', '#DC143C'], autopct='%1.1f%%', startangle=90)
ax2.set_title('Porcentaje de Resultados', fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
````

````python
# An√°lisis por equipo
# Todos los equipos que aparecen en el dataset
todos_equipos = set(df_champs['home_team'].unique()) | set(df_champs['away_team'].unique())
print(f"üìä Total de equipos en el dataset: {len(todos_equipos)}")
print(f"Equipos: {sorted(todos_equipos)}")

# Estad√≠sticas por equipo
estadisticas_equipos = []

for equipo in todos_equipos:
    # Partidos como local
    partidos_local = df_champs[df_champs['home_team'] == equipo]
    # Partidos como visitante
    partidos_visitante = df_champs[df_champs['away_team'] == equipo]
    
    total_partidos = len(partidos_local) + len(partidos_visitante)
    goles_favor = partidos_local['home_goals'].sum() + partidos_visitante['away_goals'].sum()
    goles_contra = partidos_local['away_goals'].sum() + partidos_visitante['home_goals'].sum()
    
    estadisticas_equipos.append({
        'equipo': equipo,
        'partidos': total_partidos,
        'goles_favor': goles_favor,
        'goles_contra': goles_contra,
        'diferencia_goles': goles_favor - goles_contra
    })

df_estadisticas = pd.DataFrame(estadisticas_equipos)
df_estadisticas = df_estadisticas.sort_values('diferencia_goles', ascending=False)

print("\nüèÜ Top 5 equipos por diferencia de goles:")
print(df_estadisticas.head())

# Visualizaci√≥n de estad√≠sticas de equipos con seaborn
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Gr√°fico 1: Top 10 equipos por goles a favor
top_goles = df_estadisticas.nlargest(10, 'goles_favor')
sns.barplot(data=top_goles, x='goles_favor', y='equipo', 
            palette='viridis', ax=axes[0,0])
axes[0,0].set_title('Top 10 Equipos: Goles a Favor', fontsize=12, fontweight='bold')
axes[0,0].set_xlabel('Goles a Favor')

# Gr√°fico 2: Diferencia de goles (mejores y peores)
sns.barplot(data=df_estadisticas.head(10), x='diferencia_goles', y='equipo', 
            palette='RdYlGn', ax=axes[0,1])
axes[0,1].set_title('Top 10 Equipos: Diferencia de Goles', fontsize=12, fontweight='bold')
axes[0,1].set_xlabel('Diferencia de Goles')

# Gr√°fico 3: Relaci√≥n goles a favor vs goles en contra
sns.scatterplot(data=df_estadisticas, x='goles_favor', y='goles_contra', 
                size='partidos', sizes=(50, 200), alpha=0.7, ax=axes[1,0])
axes[1,0].set_title('Goles a Favor vs Goles en Contra', fontsize=12, fontweight='bold')
axes[1,0].set_xlabel('Goles a Favor')
axes[1,0].set_ylabel('Goles en Contra')

# Gr√°fico 4: Distribuci√≥n de partidos jugados
sns.histplot(data=df_estadisticas, x='partidos', bins=10, kde=True, ax=axes[1,1])
axes[1,1].set_title('Distribuci√≥n de Partidos por Equipo', fontsize=12, fontweight='bold')
axes[1,1].set_xlabel('N√∫mero de Partidos')
axes[1,1].set_ylabel('Frecuencia')

plt.tight_layout()
plt.show()
````

````python
# Visualizaciones avanzadas con seaborn
# Agregar columna de resultado para an√°lisis m√°s profundo
def determinar_resultado(row):
    if row['home_goals'] > row['away_goals']:
        return 'Victoria Local'
    elif row['home_goals'] < row['away_goals']:
        return 'Victoria Visitante'
    else:
        return 'Empate'

df_champs['resultado'] = df_champs.apply(determinar_resultado, axis=1)

# Crear visualizaciones m√°s sofisticadas
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# 1. Boxplot de goles por fase del torneo
sns.boxplot(data=df_champs, x='stage', y='total_goals', ax=axes[0,0])
axes[0,0].set_title('Distribuci√≥n de Goles por Fase del Torneo', fontsize=12, fontweight='bold')
axes[0,0].tick_params(axis='x', rotation=45)
axes[0,0].set_xlabel('Fase del Torneo')
axes[0,0].set_ylabel('Goles Totales')

# 2. Heatmap de correlaciones
correlaciones = df_champs[['home_goals', 'away_goals', 'total_goals', 'home_possession', 'away_possession']].corr()
sns.heatmap(correlaciones, annot=True, cmap='coolwarm', center=0, ax=axes[0,1])
axes[0,1].set_title('Matriz de Correlaciones', fontsize=12, fontweight='bold')

# 3. Violinplot de posesi√≥n por resultado
df_possession_melted = pd.melt(df_champs, 
                              id_vars=['resultado'], 
                              value_vars=['home_possession', 'away_possession'],
                              var_name='tipo_posesion', 
                              value_name='posesion')
sns.violinplot(data=df_possession_melted, x='resultado', y='posesion', 
               hue='tipo_posesion', ax=axes[1,0])
axes[1,0].set_title('Distribuci√≥n de Posesi√≥n por Resultado', fontsize=12, fontweight='bold')
axes[1,0].tick_params(axis='x', rotation=45)
axes[1,0].set_xlabel('Resultado')
axes[1,0].set_ylabel('Posesi√≥n (%)')

# 4. Stripplot de goles por temporada
sns.stripplot(data=df_champs, x='season', y='total_goals', 
              size=8, alpha=0.7, ax=axes[1,1])
sns.boxplot(data=df_champs, x='season', y='total_goals', 
            ax=axes[1,1], showcaps=False, boxprops={'facecolor':'None'})
axes[1,1].set_title('Goles por Temporada (Distribuci√≥n Detallada)', fontsize=12, fontweight='bold')
axes[1,1].set_xlabel('Temporada')
axes[1,1].set_ylabel('Goles Totales')

plt.tight_layout()
plt.show()

print("üé® Visualizaciones avanzadas creadas con seaborn")
print("üí° Estas gr√°ficas muestran patrones m√°s profundos en los datos")
````
