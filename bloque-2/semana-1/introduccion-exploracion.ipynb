## 2. Preparando Nuestras Herramientas

Necesitamos importar las herramientas para trabajar con datos de f√∫tbol.

**Lo que vamos a usar:**
- üìä **pandas**: Para leer datos como si fuera Excel
- üé® **seaborn**: Para hacer gr√°ficos s√∫per bonitos

```python
# Importar las herramientas principales
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Configurar seaborn para gr√°ficos bonitos
sns.set_theme(style="whitegrid", palette="viridis")

print("‚úÖ ¬°Herramientas listas!")
print("üìä Pandas: Para leer datos de f√∫tbol")
print("üé® Seaborn: Para gr√°ficos incre√≠bles")
print("‚öΩ ¬°Listos para explorar datos deportivos!")
```

### ¬øQu√© es Pandas?

Pandas es como Excel, pero m√°s poderoso. Nos ayuda a:
- üìñ Leer datos de archivos
- üîç Buscar informaci√≥n espec√≠fica
- üßÆ Hacer c√°lculos autom√°ticamente

**Ejemplo simple:** Una tabla con equipos y sus goles

```python
# Una lista simple de equipos favoritos
equipos_favoritos = ['Barcelona', 'Real Madrid', 'Manchester City', 'Bayern Munich']

print("‚öΩ Mis equipos favoritos:")
for i, equipo in enumerate(equipos_favoritos, 1):
    print(f"{i}. {equipo}")

print(f"\nüî¢ Total de equipos: {len(equipos_favoritos)}")
```

```python
# Crear una tabla simple con datos de equipos
datos_equipos = {
    'Equipo': ['Barcelona', 'Real Madrid', 'Manchester City'],
    'Goles': [68, 87, 95],
    'Partidos': [38, 38, 38]
}

tabla_equipos = pd.DataFrame(datos_equipos)
print("üìä Tabla de equipos:")
print(tabla_equipos)

print(f"\n‚ú® ¬°Esta tabla tiene {len(tabla_equipos)} equipos y {len(tabla_equipos.columns)} columnas!")
```

## 3. ¬°Datos Reales de F√∫tbol!

Ahora vamos a trabajar con datos reales de partidos.

**¬øQu√© contienen estos datos?**
- üèüÔ∏è Equipos que jugaron
- ‚öΩ Goles marcados
- üìÖ Cu√°ndo se jug√≥
- üèÜ En qu√© competici√≥n

¬°Vamos a crear datos de ejemplo que parecen reales!

```python
# Crear datos simples de partidos de f√∫tbol
import random

# Lista de equipos famosos
equipos = ['Barcelona', 'Real Madrid', 'Manchester City', 'Bayern Munich', 
          'Liverpool', 'Paris Saint-Germain', 'Chelsea', 'Juventus']

# Crear una lista de partidos simulados
partidos = []
for i in range(30):  # Crear 30 partidos
    equipo_local = random.choice(equipos)
    equipo_visitante = random.choice([e for e in equipos if e != equipo_local])
    
    partidos.append({
        'Equipo_Local': equipo_local,
        'Equipo_Visitante': equipo_visitante,
        'Goles_Local': random.randint(0, 3),
        'Goles_Visitante': random.randint(0, 3),
        'Temporada': random.choice(['2023-24', '2024-25'])
    })

# Convertir a tabla de pandas
datos_futbol = pd.DataFrame(partidos)
print("‚öΩ ¬°Datos de f√∫tbol creados!")
print(f"üìä Tenemos {len(datos_futbol)} partidos")
print(f"üèüÔ∏è Con {len(equipos)} equipos diferentes")
```

### Echemos un Vistazo a Nuestros Datos

¬°Vamos a ver qu√© partidos tenemos!

```python
# Ver los primeros 8 partidos
print("üîç Los primeros 8 partidos:")
print(datos_futbol.head(8))

print(f"\nüìà Informaci√≥n b√°sica:")
print(f"‚Ä¢ Total de partidos: {len(datos_futbol)}")
print(f"‚Ä¢ Informaci√≥n que tenemos: {list(datos_futbol.columns)}")
```

```python
# Calcular algunas estad√≠sticas simples
total_goles = datos_futbol['Goles_Local'].sum() + datos_futbol['Goles_Visitante'].sum()
promedio_goles = total_goles / len(datos_futbol)

print("‚öΩ Estad√≠sticas B√°sicas:")
print(f"‚Ä¢ Total de goles en todos los partidos: {total_goles}")
print(f"‚Ä¢ Promedio de goles por partido: {promedio_goles:.1f}")
print(f"‚Ä¢ Partido con m√°s goles (local): {datos_futbol['Goles_Local'].max()}")
print(f"‚Ä¢ Partido con m√°s goles (visitante): {datos_futbol['Goles_Visitante'].max()}")
```

## 4. ¬°Hagamos Gr√°ficos Bonitos con Seaborn!

Ahora viene lo divertido: convertir n√∫meros en gr√°ficos incre√≠bles.

```python
# Gr√°fico 1: ¬øCu√°ntos goles marcan los equipos locales?
plt.figure(figsize=(10, 6))
sns.countplot(data=datos_futbol, x='Goles_Local', palette='viridis')
plt.title('¬øCu√°ntos Goles Marcan los Equipos Locales?', fontsize=16, fontweight='bold')
plt.xlabel('Goles Marcados')
plt.ylabel('N√∫mero de Partidos')
plt.show()

print("üìä ¬°Este gr√°fico nos muestra qu√© tan seguido los equipos marcan 0, 1, 2, 3 o 4 goles en casa!")
```

```python
# Gr√°fico 2: ¬øEs mejor jugar en casa?
datos_futbol['Total_Goles'] = datos_futbol['Goles_Local'] + datos_futbol['Goles_Visitante']

plt.figure(figsize=(10, 6))
sns.boxplot(data=datos_futbol, y='Total_Goles', palette='Set2')
plt.title('¬øCu√°ntos Goles Hay por Partido?', fontsize=16, fontweight='bold')
plt.ylabel('Total de Goles en el Partido')
plt.show()

print("üìà ¬°Este gr√°fico nos muestra si hay partidos con muchos o pocos goles!")
print(f"üéØ Promedio de goles por partido: {datos_futbol['Total_Goles'].mean():.1f}")
```

```python
# Gr√°fico 3: ¬øQu√© equipos juegan m√°s partidos?
equipos_conteo = datos_futbol['Equipo_Local'].value_counts()

plt.figure(figsize=(12, 6))
sns.barplot(x=equipos_conteo.values, y=equipos_conteo.index, palette='plasma')
plt.title('¬øQu√© Equipos Aparecen M√°s Como Locales?', fontsize=16, fontweight='bold')
plt.xlabel('N√∫mero de Partidos como Local')
plt.ylabel('Equipos')
plt.show()

print("üèüÔ∏è ¬°Este gr√°fico muestra qu√© equipos jugaron m√°s partidos en casa!")
print(f"üéØ El equipo que m√°s jug√≥ en casa: {equipos_conteo.index[0]} con {equipos_conteo.iloc[0]} partidos")
```

```python
# ¬øEs mejor jugar en casa? ¬°Vamos a averiguarlo!
# Crear una nueva columna que diga qui√©n gan√≥
def quien_gano(fila):
    if fila['Goles_Local'] > fila['Goles_Visitante']:
        return 'Gan√≥ Local'
    elif fila['Goles_Local'] < fila['Goles_Visitante']:
        return 'Gan√≥ Visitante'
    else:
        return 'Empate'

datos_futbol['Resultado'] = datos_futbol.apply(quien_gano, axis=1)

# Hacer gr√°fico con seaborn
plt.figure(figsize=(10, 6))
sns.countplot(data=datos_futbol, x='Resultado', palette='Set1')
plt.title('¬øEs Mejor Jugar en Casa?', fontsize=16, fontweight='bold')
plt.xlabel('Resultado del Partido')
plt.ylabel('N√∫mero de Partidos')
plt.show()

# Mostrar porcentajes simples
resultados = datos_futbol['Resultado'].value_counts()
total = len(datos_futbol)
print("üèüÔ∏è Resultados de nuestros partidos:")
for resultado, cantidad in resultados.items():
    porcentaje = (cantidad / total) * 100
    print(f"‚Ä¢ {resultado}: {cantidad} partidos ({porcentaje:.1f}%)")

print(f"\n‚ú® ¬°Analizamos {total} partidos en total!")
```
