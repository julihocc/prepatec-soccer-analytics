````python
# Construcción de Features Estratégicas
# Simulando las métricas que Real Madrid usa para evaluar oponentes

# Número de equipos simulados (representativo de liga mayor)
n_equipos = 500

# Métricas de Performance - Núcleo del análisis predictivo
goles_favor = np.random.normal(1.8, 0.5, n_equipos)  # Promedio goles por partido
goles_contra = np.random.normal(1.2, 0.4, n_equipos)  # Promedio goles recibidos
eficiencia_ofensiva = np.random.normal(0.15, 0.05, n_equipos)  # % de ataques efectivos

# Métricas de Consistencia - Predictores clave del éxito
consistencia_resultados = np.random.normal(0.6, 0.15, n_equipos)  # Variabilidad en resultados
rendimiento_local = np.random.normal(0.65, 0.1, n_equipos)  # % victoria como local
rendimiento_visitante = np.random.normal(0.45, 0.1, n_equipos)  # % victoria como visitante

# Métricas de Inteligencia Competitiva
experiencia_competicion = np.random.randint(0, 15, n_equipos)  # Años en competición élite
valor_plantel = np.random.lognormal(15, 1, n_equipos)  # Valor mercado (millones)
forma_reciente = np.random.normal(0.5, 0.2, n_equipos)  # % victorias últimos 5 partidos

# Variable objetivo: Éxito deportivo (lo que queremos predecir)
# Algoritmo empresarial que combina múltiples factores
probabilidad_exito = (
    0.3 * (goles_favor - goles_contra) +  # Diferencia goles
    0.25 * eficiencia_ofensiva +  # Eficiencia ofensiva
    0.2 * consistencia_resultados +  # Consistencia
    0.15 * forma_reciente +  # Forma reciente
    0.1 * (experiencia_competicion / 15)  # Experiencia normalizada
)

# Determinación estratégica del éxito (umbral empresarial)
exito_deportivo = (probabilidad_exito > np.percentile(probabilidad_exito, 60)).astype(int)

print("Features estratégicas creadas con distribuciones realistas:")
print(f"- Equipos analizados: {n_equipos}")
print(f"- Goles favor promedio: {goles_favor.mean():.2f}")
print(f"- Eficiencia ofensiva promedio: {eficiencia_ofensiva.mean():.2%}")
print(f"- Distribución de éxito: {exito_deportivo.mean():.1%} exitosos")
print()
````

## Estrategia de División de Datos en Entornos Empresariales

### ¿Por qué la división train/test es crítica en el negocio deportivo?

**Un arquitecto de inteligencia predictiva** entiende que la división de datos no es solo "técnica" - es **estrategia empresarial** que determina si un modelo vale millones o es inútil en producción.

### Principios Empresariales de División de Datos

**1. Simulación de Futuro Real**
- Train: Datos "pasados" conocidos por scouts
- Test: Datos "futuros" que el modelo debe predecir
- Proporción 80/20: Estándar para datasets empresariales

**2. Prevención de Data Leakage**
- Información del futuro no puede "contaminar" el entrenamiento
- En deportes: No usar resultados futuros para predecir presentes
- Validación cruzada temporal para series temporales

**3. Representatividad Estratégica**
- Test set debe representar condiciones reales de uso
- Balanceo similar entre train y test
- Distribución de features consistente

### La División Como Auditoría Empresarial

**Test set = Simulacro de auditoría externa**
- Ningún algoritmo ve estos datos durante entrenamiento
- Performance aquí determina confianza empresarial
- Métricas de test = ROI potencial en decisiones reales

````python
# División Estratégica de Datos para Entornos Empresariales
# Protocolo estándar usado por organizaciones deportivas élite

# Separación de features (X) y variable objetivo (y)
features_estrategicas = ['goles_favor', 'goles_contra', 'eficiencia_ofensiva', 
                        'consistencia_resultados', 'rendimiento_local', 'rendimiento_visitante',
                        'experiencia_competicion', 'valor_plantel_millones', 'forma_reciente']

X = datos_equipos[features_estrategicas]
y = datos_equipos['exito_deportivo']

# División empresarial: 80% entrenamiento, 20% validación
# Estratificación para mantener proporción de clases
X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.2,      # 20% para simulación de producción
    random_state=42,    # Reproducibilidad empresarial
    stratify=y         # Mantener distribución de éxito en ambos sets
)

print("=== ARQUITECTURA DE DIVISIÓN EMPRESARIAL ===")
print(f"Dataset total: {len(datos_equipos)} equipos analizados")
print()
print("DATOS DE ENTRENAMIENTO (Inteligencia histórica):")
print(f"  • Equipos: {len(X_train)}")
print(f"  • Features: {len(features_estrategicas)}")
print(f"  • Distribución de éxito: {y_train.mean():.1%}")
print()
print("DATOS DE VALIDACIÓN (Simulación de producción):")
print(f"  • Equipos: {len(X_test)}")
print(f"  • Features: {len(features_estrategicas)}")
print(f"  • Distribución de éxito: {y_test.mean():.1%}")
print()

print("=== VALIDACIÓN DE DIVISIÓN ESTRATÉGICA ===")
print("Verificando representatividad entre sets:")
print("\nComparación de medias (Train vs Test):")
for feature in features_estrategicas:
    media_train = X_train[feature].mean()
    media_test = X_test[feature].mean()
    diferencia = abs(media_train - media_test) / media_train * 100
    print(f"  {feature}: {diferencia:.1f}% diferencia")

print("\nDivisión estratégica completada - Lista para entrenamiento empresarial")
````

## Marco Estratégico de Algoritmos en Inteligencia Deportiva

### ¿Cómo seleccionar algoritmos para decisiones de millones de euros?

**Un arquitecto de inteligencia predictiva** no elige algoritmos por "popularidad" o porque "funcionan bien". Selecciona **herramientas algorítmicas** basándose en requisitos empresariales específicos y contexto estratégico.

### Portfolio de Algoritmos Empresariales

**1. Regresión Logística: El "Ejecutivo Senior"**
- **Fortalezas:** Interpretabilidad total, coeficientes explicables
- **Caso de uso:** Presentaciones a directivos, auditorías, decisiones regulatorias
- **Cuándo usarlo:** Cuando necesitas explicar "por qué" el modelo decidió algo

**2. Random Forest: El "Consultor Especialista"**
- **Fortalezas:** Maneja interacciones complejas, robusto a outliers
- **Caso de uso:** Análisis profundo, captura patrones no lineales
- **Cuándo usarlo:** Cuando la precisión supera la interpretabilidad simple

### Marco de Decisión Algorítmica

**Pregunta empresarial fundamental:**
¿Qué es más valioso para tu organización: explicar cada decisión o maximizar la precisión predictiva?

**Contexto Real Madrid vs Athletic Bilbao:**
- **Regresión Logística:** "Tenemos 75% probabilidad de ganar porque nuestros goles a favor son 1.2 superiores al promedio"
- **Random Forest:** "Tenemos 78% probabilidad de ganar (basado en 127 interacciones complejas)"

### Consideraciones de Implementación Empresarial

**1. Escalabilidad:** ¿Funciona con 50,000 equipos?
**2. Velocidad:** ¿Predice en tiempo real durante transferencias?
**3. Mantenimiento:** ¿Puede ser actualizado por el equipo técnico?
**4. Regulación:** ¿Cumple con requisitos de transparencia?

### Reflexión Estratégica

**Si fueras contratado por la UEFA para evaluar sistemas predictivos:**
- ¿Qué algoritmo recomendarías para decisiones de Fair Play Financiero?
- ¿Cómo balancearías precisión vs explicabilidad?
- ¿Qué criterios definirían un "sistema confiable"?

````python
# Entrenamiento de Portfolio Algorítmico Empresarial
# Implementación de múltiples estrategias para comparación ejecutiva

print("=== LABORATORIO DE ALGORITMOS EMPRESARIALES ===")
print("Entrenando portfolio de modelos para análisis comparativo...")
print()

# MODELO 1: Regresión Logística - "El Ejecutivo Senior"
# Interpretable, explicable, auditable
modelo_regresion = LogisticRegression(random_state=42, max_iter=1000)
modelo_regresion.fit(X_train, y_train)

# MODELO 2: Random Forest - "El Consultor Especialista"  
# Potente, robusto, captura patrones complejos
modelo_forest = RandomForestClassifier(
    n_estimators=100,    # 100 árboles para estabilidad
    random_state=42,     # Reproducibilidad empresarial
    max_depth=10         # Control de overfitting
)
modelo_forest.fit(X_train, y_train)

print("ENTRENAMIENTO COMPLETADO:")
print("✓ Regresión Logística: Modelo interpretable entrenado")
print("✓ Random Forest: Modelo de ensemble entrenado")
print()

# Predicciones en datos de validación (simulación de producción)
predicciones_regresion = modelo_regresion.predict(X_test)
predicciones_forest = modelo_forest.predict(X_test)

# Métricas empresariales de rendimiento
accuracy_regresion = accuracy_score(y_test, predicciones_regresion)
accuracy_forest = accuracy_score(y_test, predicciones_forest)

print("=== MÉTRICAS EJECUTIVAS DE RENDIMIENTO ===")
print(f"Regresión Logística - Precisión: {accuracy_regresion:.1%}")
print(f"Random Forest - Precisión: {accuracy_forest:.1%}")
print(f"Diferencia de rendimiento: {abs(accuracy_forest - accuracy_regresion):.1%}")
print()

# Análisis de importancia de features (inteligencia empresarial)
print("=== INTELIGENCIA DE FEATURES ===")
print("Importancia según Random Forest (Top 5 factores predictivos):")
importancias = pd.DataFrame({
    'feature': features_estrategicas,
    'importancia': modelo_forest.feature_importances_
}).sort_values('importancia', ascending=False)

for i, (feature, importancia) in enumerate(importancias.head().values):
    print(f"{i+1}. {feature}: {importancia:.1%}")

print()
print("Modelos entrenados y evaluados - Listos para análisis empresarial")
````

````python
# Simulación Empresarial: Evaluación de Transferencia en Tiempo Real
# Caso práctico: Manchester City evaluando adquisición estratégica

print("=== LABORATORIO DE DECISIONES EMPRESARIALES ===")
print("Simulando evaluación de transferencia con inteligencia predictiva...")
print()

# PERFIL DEL EQUIPO CANDIDATO (Datos reales del departamento de scouting)
equipo_candidato = pd.DataFrame({
    'goles_favor': [1.6],                    # Ligeramente por debajo del promedio
    'goles_contra': [1.1],                   # Defensa sólida
    'eficiencia_ofensiva': [0.13],           # Eficiencia moderada
    'consistencia_resultados': [0.45],       # Inconsistencia notable
    'rendimiento_local': [0.70],             # Fuerte como local
    'rendimiento_visitante': [0.35],         # Débil como visitante
    'experiencia_competicion': [3],          # Experiencia limitada
    'valor_plantel_millones': [45.0],        # Valor moderado
    'forma_reciente': [0.40]                 # Forma irregular reciente
})

print("PERFIL EMPRESARIAL DEL CANDIDATO:")
print("=" * 40)
for feature, valor in equipo_candidato.iloc[0].items():
    if feature in ['goles_favor', 'goles_contra', 'eficiencia_ofensiva']:
        print(f"{feature:25}: {valor:.2f}")
    elif feature in ['valor_plantel_millones']:
        print(f"{feature:25}: €{valor:.1f}M")
    elif feature in ['experiencia_competicion']:
        print(f"{feature:25}: {valor} años")
    else:
        print(f"{feature:25}: {valor:.1%}")
print()

# PREDICCIONES EMPRESARIALES CON AMBOS MODELOS
prob_regresion = modelo_regresion.predict_proba(equipo_candidato)[0]
prob_forest = modelo_forest.predict_proba(equipo_candidato)[0]

prediccion_regresion = modelo_regresion.predict(equipo_candidato)[0]
prediccion_forest = modelo_forest.predict(equipo_candidato)[0]

print("=== ANÁLISIS PREDICTIVO EMPRESARIAL ===")
print()
print("1. REGRESIÓN LOGÍSTICA (Análisis Interpretable):")
print(f"   • Probabilidad de éxito: {prob_regresion[1]:.1%}")
print(f"   • Recomendación: {'ADQUIRIR' if prediccion_regresion == 1 else 'NO ADQUIRIR'}")
print(f"   • Nivel de confianza: {max(prob_regresion):.1%}")
print()
print("2. RANDOM FOREST (Análisis Avanzado):")
print(f"   • Probabilidad de éxito: {prob_forest[1]:.1%}")
print(f"   • Recomendación: {'ADQUIRIR' if prediccion_forest == 1 else 'NO ADQUIRIR'}")
print(f"   • Nivel de confianza: {max(prob_forest):.1%}")
print()

# ANÁLISIS DE FACTORES CRÍTICOS (Justificación técnica)
print("=== ANÁLISIS DE FACTORES CRÍTICOS ===")
print("Justificación técnica para la recomendación:")
print()

# Comparar con promedio del dataset para contexto
print("BENCHMARKING vs MERCADO:")
for feature in features_estrategicas:
    valor_candidato = equipo_candidato[feature].iloc[0]
    promedio_mercado = X[feature].mean()
    diferencia = ((valor_candidato - promedio_mercado) / promedio_mercado) * 100
    
    estado = "SUPERIOR" if diferencia > 5 else "INFERIOR" if diferencia < -5 else "PROMEDIO"
    print(f"• {feature:25}: {diferencia:+5.1f}% vs mercado ({estado})")

print()

# RECOMENDACIÓN EJECUTIVA FINAL
consensus = "ADQUIRIR" if (prediccion_regresion == 1 and prediccion_forest == 1) else \
           "NO ADQUIRIR" if (prediccion_regresion == 0 and prediccion_forest == 0) else \
           "ANÁLISIS ADICIONAL REQUERIDO"

print("=" * 50)
print("=== RECOMENDACIÓN EJECUTIVA FINAL ===")
print(f"DECISIÓN EMPRESARIAL: {consensus}")
print(f"CONSENSO DE MODELOS: {consensus != 'ANÁLISIS ADICIONAL REQUERIDO'}")
print(f"PROBABILIDAD PROMEDIO: {(prob_regresion[1] + prob_forest[1])/2:.1%}")
print()

if consensus == "ADQUIRIR":
    print("JUSTIFICACIÓN: Ambos modelos recomiendan la adquisición")
    print("PRÓXIMOS PASOS: Proceder con due diligence detallada")
elif consensus == "NO ADQUIRIR":
    print("JUSTIFICACIÓN: Ambos modelos desaconsejan la adquisición")
    print("PRÓXIMOS PASOS: Buscar alternativas en el mercado")
else:
    print("JUSTIFICACIÓN: Modelos en desacuerdo - requiere análisis humano")
    print("PRÓXIMOS PASOS: Evaluación adicional con comité técnico")

print()
print("Análisis empresarial completado - Decisión lista para board directivo")
````
