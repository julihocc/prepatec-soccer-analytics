## 3. Datos de F√∫tbol para Practicar

### 3.1 Creaci√≥n de Datos Simples de Champions League

Vamos a crear datos que simulen partidos reales de la Champions League:

- **Temporadas**: 2018-2019 a 2022-2023 (5 temporadas)
- **Equipos**: Los equipos m√°s famosos de Europa
- **Estad√≠sticas**: Goles, posesi√≥n, tiros, pases
- **Fases**: Grupos, octavos, cuartos, semifinales, final

### ¬øPor qu√© Crear Datos Falsos?
- Para practicar sin problemas de datos reales
- Para tener control sobre lo que analizamos
- Para aprender sin complicaciones t√©cnicas
- Para enfocarnos en el an√°lisis, no en conseguir datos

```python
# Crear dataset de f√∫tbol para practicar
print("GENERANDO DATOS DE CHAMPIONS LEAGUE")
print("=" * 40)

# Equipos famosos que participan
equipos = [
    'Real Madrid', 'Barcelona', 'Manchester City', 'Liverpool',
    'Bayern Munich', 'Paris Saint-Germain', 'Chelsea', 'Manchester United',
    'Atletico Madrid', 'Juventus', 'Inter Milan', 'AC Milan',
    'Arsenal', 'Tottenham', 'Borussia Dortmund', 'Sevilla'
]

# Temporadas que vamos a analizar
temporadas = ['2018-19', '2019-20', '2020-21', '2021-22', '2022-23']

# Fases del torneo
fases = ['Grupos', 'Octavos', 'Cuartos', 'Semifinales', 'Final']

# Generar partidos
partidos_data = []
partido_id = 1

for temporada in temporadas:
    print(f"Generando temporada {temporada}...")
    
    for fase in fases:
        # Cu√°ntos partidos por fase
        if fase == 'Grupos':
            n_partidos = 48  # Simplificado
        elif fase == 'Octavos':
            n_partidos = 16
        elif fase == 'Cuartos':
            n_partidos = 8
        elif fase == 'Semifinales':
            n_partidos = 4
        else:  # Final
            n_partidos = 1
            
        for i in range(n_partidos):
            # Elegir dos equipos diferentes
            equipos_partido = np.random.choice(equipos, 2, replace=False)
            equipo_local = equipos_partido[0]
            equipo_visitante = equipos_partido[1]
            
            # Simular goles (los locales tienen peque√±a ventaja)
            goles_local = np.random.randint(0, 4)
            goles_visitante = np.random.randint(0, 3)
            
            # Quien gan√≥
            if goles_local > goles_visitante:
                resultado = 'Local'
            elif goles_visitante > goles_local:
                resultado = 'Visitante'
            else:
                resultado = 'Empate'
            
            # Otras estad√≠sticas simples
            posesion_local = np.random.randint(35, 65)
            tiros_local = np.random.randint(5, 20)
            tiros_visitante = np.random.randint(5, 20)
            
            # Guardar el partido
            partidos_data.append({
                'partido_id': partido_id,
                'temporada': temporada,
                'fase': fase,
                'equipo_local': equipo_local,
                'equipo_visitante': equipo_visitante,
                'goles_local': goles_local,
                'goles_visitante': goles_visitante,
                'resultado': resultado,
                'posesion_local': posesion_local,
                'tiros_local': tiros_local,
                'tiros_visitante': tiros_visitante,
                'total_goles': goles_local + goles_visitante
            })
            
            partido_id += 1

# Crear tabla de datos
df_champions = pd.DataFrame(partidos_data)

# Ver qu√© creamos
print(f"\nDATOS CREADOS EXITOSAMENTE")
print("=" * 40)
print(f"Total de partidos: {len(df_champions)}")
print(f"Temporadas: {len(temporadas)}")
print(f"Equipos: {len(equipos)}")
print(f"Fases: {len(fases)}")

# Estad√≠sticas b√°sicas
print(f"\nESTAD√çSTICAS B√ÅSICAS:")
print(f"- Promedio de goles por partido: {df_champions['total_goles'].mean():.2f}")
print(f"- M√°ximo goles en un partido: {df_champions['total_goles'].max()}")
print(f"- Partidos con m√°s de 3 goles: {(df_champions['total_goles'] > 3).sum()}")

# Ver algunos partidos
print(f"\nEJEMPLOS DE PARTIDOS:")
print(df_champions[['temporada', 'equipo_local', 'equipo_visitante', 'goles_local', 'goles_visitante', 'resultado']].head(8))
```

## 4. Nuestro Primer An√°lisis

### 4.1 Analizar una Temporada Espec√≠fica

Vamos a elegir una temporada y analizarla a fondo. Esto nos ayudar√° a entender c√≥mo hacer un an√°lisis paso a paso.

```python
# Analizar la temporada 2021-22
print("AN√ÅLISIS DE UNA TEMPORADA: 2021-22")
print("=" * 40)

# Filtrar solo los datos de esa temporada
temporada_elegida = '2021-22'
df_temporada = df_champions[df_champions['temporada'] == temporada_elegida].copy()

print(f"DATOS DE LA TEMPORADA {temporada_elegida}")
print(f"- Total de partidos: {len(df_temporada)}")
print(f"- Equipos que participaron: {df_temporada['equipo_local'].nunique()}")

# Estad√≠sticas b√°sicas de goles
print(f"\nAN√ÅLISIS DE GOLES")
print("-" * 30)
goles_stats = df_temporada['total_goles'].describe()
print(f"Promedio: {goles_stats['mean']:.2f} goles por partido")
print(f"M√≠nimo: {goles_stats['min']:.0f} goles")
print(f"M√°ximo: {goles_stats['max']:.0f} goles")
print(f"Mediana: {goles_stats['50%']:.2f} goles")

# ¬øQu√© significa esto?
print(f"\n¬øQU√â SIGNIFICA?")
print("-" * 30)
if goles_stats['mean'] > 2.5:
    print("Esta temporada fue MUY goleadora")
elif goles_stats['mean'] > 2.0:
    print("Esta temporada tuvo buenos goles")
else:
    print("Esta temporada fue defensiva")

# An√°lisis por fases
print(f"\nAN√ÅLISIS POR FASES")
print("-" * 30)
goles_por_fase = df_temporada.groupby('fase')['total_goles'].agg(['mean', 'count']).round(2)
goles_por_fase.columns = ['Goles_Promedio', 'Partidos']
print(goles_por_fase)

# Encontrar la fase m√°s goleadora
fase_mas_goles = goles_por_fase['Goles_Promedio'].idxmax()
print(f"\nLa fase m√°s goleadora fue: {fase_mas_goles}")

# An√°lisis de ventaja local
print(f"\nVENTAJA LOCAL")
print("-" * 30)
resultados = df_temporada['resultado'].value_counts()
total_partidos = len(df_temporada)

print(f"Victorias locales: {resultados['Local']} ({resultados['Local']/total_partidos:.1%})")
print(f"Victorias visitantes: {resultados['Visitante']} ({resultados['Visitante']/total_partidos:.1%})")
print(f"Empates: {resultados['Empate']} ({resultados['Empate']/total_partidos:.1%})")

# ¬øHay ventaja local?
if resultados['Local'] > resultados['Visitante']:
    print("\n¬°S√ç hay ventaja local!")
    diferencia = resultados['Local'] - resultados['Visitante']
    print(f"Los locales ganaron {diferencia} partidos m√°s")
else:
    print("\nNo hay ventaja local clara")

# Crear gr√°fico simple con seaborn
plt.figure(figsize=(12, 8))

# Gr√°fico 1: Distribuci√≥n de goles
plt.subplot(2, 2, 1)
sns.histplot(data=df_temporada, x='total_goles', bins=8)
plt.title('Distribuci√≥n de Goles por Partido')

# Gr√°fico 2: Goles por fase
plt.subplot(2, 2, 2)
sns.barplot(data=df_temporada, x='fase', y='total_goles')
plt.title('Goles Promedio por Fase')
plt.xticks(rotation=45)

# Gr√°fico 3: Resultados
plt.subplot(2, 2, 3)
resultado_counts = df_temporada['resultado'].value_counts()
plt.pie(resultado_counts.values, labels=resultado_counts.index, autopct='%1.1f%%')
plt.title('Distribuci√≥n de Resultados')

# Gr√°fico 4: Posesi√≥n vs Goles locales
plt.subplot(2, 2, 4)
sns.scatterplot(data=df_temporada, x='posesion_local', y='goles_local')
plt.title('Posesi√≥n vs Goles (Equipos Locales)')

plt.tight_layout()
plt.show()

print(f"\nCONCLUSIONES DE LA TEMPORADA {temporada_elegida}:")
print("1. Los equipos locales tienen ventaja" if resultados['Local'] > resultados['Visitante'] else "1. No hay ventaja local clara")
print(f"2. La fase m√°s emocionante fue: {fase_mas_goles}")
print(f"3. En promedio hubo {goles_stats['mean']:.1f} goles por partido")
```

## 5. Comparar Diferentes Temporadas

### 5.1 ¬øHan Cambiado los Goles con el Tiempo?

Ahora vamos a comparar todas las temporadas para ver si hay diferencias entre ellas. ¬øEl f√∫tbol ha cambiado?

```python
# Comparar todas las temporadas
print("COMPARACI√ìN ENTRE TEMPORADAS")
print("=" * 40)

# Calcular estad√≠sticas por temporada
stats_temporadas = df_champions.groupby('temporada').agg({
    'total_goles': 'mean',
    'posesion_local': 'mean',
    'tiros_local': 'mean',
    'tiros_visitante': 'mean'
}).round(2)

stats_temporadas.columns = ['Goles_Promedio', 'Posesion_Local', 'Tiros_Local', 'Tiros_Visitante']

print("ESTAD√çSTICAS POR TEMPORADA:")
print(stats_temporadas)

# ¬øCu√°l temporada fue m√°s goleadora?
temp_mas_goles = stats_temporadas['Goles_Promedio'].idxmax()
temp_menos_goles = stats_temporadas['Goles_Promedio'].idxmin()

print(f"\nTemporada M√ÅS goleadora: {temp_mas_goles} ({stats_temporadas.loc[temp_mas_goles, 'Goles_Promedio']:.2f} goles/partido)")
print(f"Temporada MENOS goleadora: {temp_menos_goles} ({stats_temporadas.loc[temp_menos_goles, 'Goles_Promedio']:.2f} goles/partido)")

# Crear gr√°ficos de comparaci√≥n con seaborn
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('EVOLUCI√ìN DE LA CHAMPIONS LEAGUE', fontsize=16, fontweight='bold')

# Preparar datos para seaborn
datos_temporadas = stats_temporadas.reset_index()

# Gr√°fico 1: Evoluci√≥n de goles
sns.lineplot(data=datos_temporadas, x='temporada', y='Goles_Promedio', 
             marker='o', linewidth=3, ax=axes[0, 0])
axes[0, 0].set_title('Goles por Temporada')
axes[0, 0].set_ylabel('Goles promedio')
axes[0, 0].tick_params(axis='x', rotation=45)

# Gr√°fico 2: Posesi√≥n local
sns.lineplot(data=datos_temporadas, x='temporada', y='Posesion_Local', 
             marker='s', linewidth=3, ax=axes[0, 1])
axes[0, 1].axhline(y=50, color='red', linestyle='--', alpha=0.7)
axes[0, 1].set_title('Posesi√≥n de Equipos Locales')
axes[0, 1].set_ylabel('Posesi√≥n promedio (%)')
axes[0, 1].tick_params(axis='x', rotation=45)

# Gr√°fico 3: Tiros por temporada
sns.lineplot(data=datos_temporadas, x='temporada', y='Tiros_Local', 
             marker='o', linewidth=3, label='Local', ax=axes[1, 0])
sns.lineplot(data=datos_temporadas, x='temporada', y='Tiros_Visitante', 
             marker='s', linewidth=3, label='Visitante', ax=axes[1, 0])
axes[1, 0].set_title('Tiros por Temporada')
axes[1, 0].set_ylabel('Tiros promedio')
axes[1, 0].tick_params(axis='x', rotation=45)
axes[1, 0].legend()

# Gr√°fico 4: Distribuci√≥n de goles por temporada
sns.boxplot(data=df_champions, x='temporada', y='total_goles', ax=axes[1, 1])
axes[1, 1].set_title('Distribuci√≥n de Goles')
axes[1, 1].set_ylabel('Goles por partido')
axes[1, 1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# An√°lisis de tendencias
print(f"\nAN√ÅLISIS DE TENDENCIAS")
print("-" * 30)

# Calcular si hay tendencia en goles
goles_valores = stats_temporadas['Goles_Promedio'].values
diferencia_total = goles_valores[-1] - goles_valores[0]

print(f"Cambio en goles del 2018-19 al 2022-23: {diferencia_total:+.2f}")

if abs(diferencia_total) > 0.2:
    if diferencia_total > 0:
        print("TENDENCIA: Los goles han AUMENTADO con el tiempo")
    else:
        print("TENDENCIA: Los goles han DISMINUIDO con el tiempo")
else:
    print("TENDENCIA: Los goles se mantienen ESTABLES")

# An√°lisis de posesi√≥n
posesion_valores = stats_temporadas['Posesion_Local'].values
diferencia_posesion = posesion_valores[-1] - posesion_valores[0]

print(f"\nCambio en posesi√≥n local: {diferencia_posesion:+.1f}%")

if abs(diferencia_posesion) > 2:
    if diferencia_posesion > 0:
        print("TENDENCIA: La ventaja local ha AUMENTADO")
    else:
        print("TENDENCIA: La ventaja local ha DISMINUIDO")
else:
    print("TENDENCIA: La posesi√≥n local se mantiene ESTABLE")

print(f"\nCONCLUSIONES PRINCIPALES:")
print(f"1. La temporada m√°s emocionante: {temp_mas_goles}")
print(f"2. Los goles {'aumentaron' if diferencia_total > 0.1 else 'se mantuvieron estables'}")
print(f"3. La ventaja local {'cambi√≥' if abs(diferencia_posesion) > 2 else 'se mantuvo'}")
```

## 6. Buscar Patrones Interesantes

### 6.1 ¬øQu√© Variables Est√°n Relacionadas?

Ahora vamos a buscar patrones en nuestros datos. ¬øM√°s tiros significa m√°s goles? ¬øLa posesi√≥n es importante?

```python
# Buscar patrones en los datos
print("AN√ÅLISIS DE PATRONES")
print("=" * 30)

# Preparar datos simples para an√°lisis
# Combinar datos de equipos locales y visitantes
datos_equipos = []

for _, partido in df_champions.iterrows():
    # Datos del equipo local
    datos_equipos.append({
        'goles': partido['goles_local'],
        'posesion': partido['posesion_local'],
        'tiros': partido['tiros_local'],
        'tipo': 'Local'
    })
    
    # Datos del equipo visitante
    datos_equipos.append({
        'goles': partido['goles_visitante'],
        'posesion': 100 - partido['posesion_local'],  # Posesi√≥n visitante
        'tiros': partido['tiros_visitante'],
        'tipo': 'Visitante'
    })

df_equipos = pd.DataFrame(datos_equipos)

# Calcular correlaciones simples
correlaciones = df_equipos[['goles', 'posesion', 'tiros']].corr()

print("CORRELACIONES ENTRE VARIABLES:")
print(correlaciones.round(3))

# ¬øQu√© est√° m√°s relacionado con goles?
corr_goles = correlaciones['goles'].abs().sort_values(ascending=False)
print(f"\nVARIABLES M√ÅS RELACIONADAS CON GOLES:")
for variable, correlacion in corr_goles.items():
    if variable != 'goles':
        print(f"- {variable}: {correlacion:.3f}")

# Interpretar la correlaci√≥n m√°s fuerte
var_importante = corr_goles.index[1]  # La segunda (primera es goles consigo mismo)
corr_importante = corr_goles.iloc[1]

print(f"\nLA VARIABLE M√ÅS IMPORTANTE: {var_importante}")
if corr_importante > 0.5:
    print("Esta correlaci√≥n es FUERTE - muy importante")
elif corr_importante > 0.3:
    print("Esta correlaci√≥n es MODERADA - importante")
else:
    print("Esta correlaci√≥n es D√âBIL - poco importante")

# An√°lisis de eficiencia
print(f"\nAN√ÅLISIS DE EFICIENCIA")
print("-" * 30)

# Calcular goles por tiro (eficiencia)
df_equipos['eficiencia'] = df_equipos['goles'] / df_equipos['tiros']
# Remover casos donde no hubo tiros
df_equipos['eficiencia'] = df_equipos['eficiencia'].replace([np.inf, -np.inf], np.nan)

eficiencia_promedio = df_equipos['eficiencia'].mean()
print(f"Eficiencia promedio: {eficiencia_promedio:.3f} goles por tiro")
print(f"Esto significa que se necesitan ~{1/eficiencia_promedio:.0f} tiros para anotar un gol")

# Crear visualizaciones con seaborn
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('PATRONES EN LOS DATOS DE CHAMPIONS', fontsize=16, fontweight='bold')

# Gr√°fico 1: Tiros vs Goles
sns.scatterplot(data=df_equipos, x='tiros', y='goles', hue='posesion', ax=axes[0, 0])
sns.regplot(data=df_equipos, x='tiros', y='goles', scatter=False, color='red', ax=axes[0, 0])
axes[0, 0].set_title('¬øM√°s Tiros = M√°s Goles?')

# Gr√°fico 2: Goles por tipo de equipo
sns.boxplot(data=df_equipos, x='tipo', y='goles', ax=axes[0, 1])
axes[0, 1].set_title('Goles: Local vs Visitante')

# Gr√°fico 3: Distribuci√≥n de eficiencia
sns.histplot(data=df_equipos, x='eficiencia', bins=30, ax=axes[1, 0])
axes[1, 0].axvline(eficiencia_promedio, color='red', linestyle='--', 
                   label=f'Promedio: {eficiencia_promedio:.3f}')
axes[1, 0].set_title('Distribuci√≥n de Eficiencia')
axes[1, 0].legend()

# Gr√°fico 4: Mapa de calor de correlaciones
sns.heatmap(correlaciones, annot=True, cmap='coolwarm', center=0, ax=axes[1, 1])
axes[1, 1].set_title('Correlaciones Entre Variables')

plt.tight_layout()
plt.show()

# Comparar equipos locales vs visitantes
print(f"\nCOMPARACI√ìN LOCAL VS VISITANTE")
print("-" * 30)

goles_local = df_equipos[df_equipos['tipo'] == 'Local']['goles'].mean()
goles_visitante = df_equipos[df_equipos['tipo'] == 'Visitante']['goles'].mean()
diferencia = goles_local - goles_visitante

print(f"Goles promedio local: {goles_local:.3f}")
print(f"Goles promedio visitante: {goles_visitante:.3f}")
print(f"Diferencia: {diferencia:+.3f} goles")

if abs(diferencia) > 0.1:
    if diferencia > 0:
        print("CONCLUSI√ìN: Los equipos locales S√ç tienen ventaja")
    else:
        print("CONCLUSI√ìN: Los equipos visitantes tienen ventaja (raro)")
else:
    print("CONCLUSI√ìN: No hay diferencia significativa")

print(f"\nINSIGHTS PRINCIPALES:")
print(f"1. La variable m√°s relacionada con goles es: {var_importante}")
print(f"2. Se necesitan ~{1/eficiencia_promedio:.0f} tiros para anotar un gol")
print(f"3. Ventaja local: {'S√ç' if diferencia > 0.1 else 'NO'} existe")
print(f"4. La correlaci√≥n m√°s fuerte es {corr_importante:.3f} ({'fuerte' if corr_importante > 0.5 else 'moderada'})")
```
