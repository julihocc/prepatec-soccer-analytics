## 1. ¬øQu√© son los Gr√°ficos?

Los **gr√°ficos** son dibujos que nos ayudan a entender n√∫meros m√°s f√°cilmente.

### Un Ejemplo Simple

Imagina que quieres saber qu√© equipo marc√≥ m√°s goles:

- **Solo n√∫meros**: Barcelona: 25, Madrid: 20, Valencia: 15
- **Con gr√°fico**: Ver√≠as inmediatamente que Barcelona tiene la barra m√°s alta

### Los Gr√°ficos que Aprenderemos

1. **Gr√°ficos de barras**: Para comparar equipos (¬øqui√©n marc√≥ m√°s goles?)
2. **Gr√°ficos de l√≠neas**: Para ver c√≥mo cambian las cosas en el tiempo

¬°Es m√°s f√°cil de lo que piensas!

## 2. Preparando Python para Hacer Gr√°ficos

Antes de dibujar, necesitamos decirle a Python qu√© herramientas usar.

Vamos a usar **seaborn** porque hace gr√°ficos m√°s bonitos autom√°ticamente.

```python
# Importar las herramientas para hacer gr√°ficos bonitos
import seaborn as sns
import matplotlib.pyplot as plt

# Configurar seaborn para gr√°ficos m√°s atractivos
sns.set_theme()

# Verificar que todo funcione
print("‚úÖ ¬°Listo para hacer gr√°ficos bonitos!")
print("üé® Seaborn har√° que nuestros gr√°ficos se vean incre√≠bles")
```

## 3. Mi Primer Gr√°fico: Barras

Los **gr√°ficos de barras** son perfectos cuando queremos comparar cosas.

**¬øCu√°ndo los usamos?**
- Para ver qu√© equipo marc√≥ m√°s goles
- Para comparar cu√°ntos partidos gan√≥ cada equipo
- Para saber qu√© jugador es m√°s alto

¬°Vamos a hacer nuestro primer gr√°fico!

```python
# Datos simples: goles de algunos equipos
equipos = ['Barcelona', 'Madrid', 'Valencia']
goles = [25, 20, 15]

# Crear el gr√°fico de barras con seaborn (¬°m√°s f√°cil!)
sns.barplot(x=equipos, y=goles)

# Agregar t√≠tulo
plt.title('Goles por Equipo')

# Mostrar el gr√°fico
plt.show()

print("üéâ ¬°Hiciste tu primer gr√°fico!")
print("üìä ¬°Mira qu√© bonito se ve con seaborn!")
print("üåà Los colores son autom√°ticos y profesionales")
```

### ¬øPor qu√© usar Seaborn?

Seaborn es genial porque:
- **Colores autom√°ticos**: Ya no tienes que elegir colores
- **M√°s bonito**: Los gr√°ficos se ven profesionales sin esfuerzo
- **M√°s f√°cil**: Menos c√≥digo para el mismo resultado
- **Estilo moderno**: Como los gr√°ficos que ves en deportes profesionales

```python
# Los mismos datos de antes
equipos = ['Barcelona', 'Madrid', 'Valencia']
goles = [25, 20, 15]

# Crear gr√°fico con seaborn y una paleta de colores bonita
sns.barplot(x=equipos, y=goles, palette='viridis')

# Agregar t√≠tulo y etiquetas
plt.title('Goles por Equipo', fontsize=16)
plt.xlabel('Equipos')
plt.ylabel('N√∫mero de Goles')

# Mostrar el gr√°fico
plt.show()

print("üåà ¬°Seaborn eligi√≥ colores incre√≠bles autom√°ticamente!")
print("‚ú® La paleta 'viridis' es muy popular en ciencia de datos")
print("üéØ ¬°Con una l√≠nea de c√≥digo obtuvimos un gr√°fico profesional!")
```

## 4. Gr√°ficos de L√≠neas: Viendo Cambios en el Tiempo

Los **gr√°ficos de l√≠neas** son perfectos para ver c√≥mo cambian las cosas.

**¬øCu√°ndo los usamos?**
- Para ver si un equipo est√° mejorando mes a mes
- Para seguir cu√°ntos goles marca un jugador cada partido

¬°Vamos a hacer uno!

```python
# Datos simples: goles de Barcelona en cada mes
meses = ['Enero', 'Febrero', 'Marzo', 'Abril']
goles_barcelona = [5, 8, 6, 10]

# Crear el gr√°fico de l√≠neas con seaborn
sns.lineplot(x=meses, y=goles_barcelona, marker='o', linewidth=3, markersize=10)

# Agregar t√≠tulo y etiquetas
plt.title('Goles de Barcelona por Mes')
plt.xlabel('Mes')
plt.ylabel('Goles')

# Mostrar el gr√°fico
plt.show()

print("üìà ¬°Mira qu√© l√≠nea tan bonita hizo seaborn!")
print("üîµ Los puntos y la l√≠nea se ven s√∫per profesionales")
print("‚ú® Seaborn hace que todo se vea mejor autom√°ticamente")
```
