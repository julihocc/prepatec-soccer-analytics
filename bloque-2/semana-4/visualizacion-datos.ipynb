## 1. Teor√≠a: Principios de Visualizaci√≥n Efectiva

### 1.1 ¬øPor qu√© Visualizar Datos?

La visualizaci√≥n de datos es **fundamental** en el an√°lisis deportivo porque:

#### **üß† Percepci√≥n Visual**
- El cerebro humano procesa informaci√≥n visual **mucho m√°s r√°pido** que texto
- Los patrones son m√°s f√°ciles de identificar en gr√°ficos que en tablas
- La memoria visual es m√°s duradera que la textual

#### **üìä Comunicaci√≥n**
- **Simplifica** conceptos complejos
- **Persuade** con evidencia visual
- **Unifica** la comprensi√≥n entre audiencias diversas

#### **‚öΩ An√°lisis Deportivo**
- **Identifica tendencias** en rendimiento
- **Compara** jugadores y equipos de forma intuitiva
- **Detecta anomal√≠as** y oportunidades

### 1.2 Principios de Visualizaci√≥n Efectiva

#### **1.2.1 Claridad y Simplicidad**
- **Una historia por gr√°fico**: Enfoque en un mensaje principal
- **Eliminar ruido visual**: Menos elementos = mayor impacto
- **Jerarqu√≠a visual**: Destacar lo m√°s importante

#### **1.2.2 Precisi√≥n e Integridad**
- **Escalas apropiadas**: Empezar ejes en cero cuando sea relevante
- **Proporciones reales**: Evitar distorsiones enga√±osas
- **Contexto completo**: Incluir informaci√≥n necesaria para interpretaci√≥n

#### **1.2.3 Est√©tica y Profesionalismo**
- **Colores significativos**: Paletas que refuercen el mensaje
- **Tipograf√≠a legible**: Tama√±os y fuentes apropiadas
- **Consistencia**: Mismo estilo en toda la presentaci√≥n

### 1.3 Tipos de Gr√°ficos y Cu√°ndo Usarlos

#### **üìà Gr√°ficos de L√≠neas**
- **Prop√≥sito**: Mostrar cambios en el tiempo
- **Ejemplo**: Evoluci√≥n de goles por temporada
- **Cu√°ndo usar**: Datos temporales, tendencias

#### **üìä Gr√°ficos de Barras**
- **Prop√≥sito**: Comparar categor√≠as
- **Ejemplo**: Goles por equipo en una temporada
- **Cu√°ndo usar**: Comparaciones entre grupos

#### **üîç Gr√°ficos de Dispersi√≥n**
- **Prop√≥sito**: Mostrar relaciones entre variables
- **Ejemplo**: Posesi√≥n vs. goles anotados
- **Cu√°ndo usar**: Correlaciones, patrones

#### **üì¶ Boxplots**
- **Prop√≥sito**: Mostrar distribuci√≥n y outliers
- **Ejemplo**: Distribuci√≥n de edades por posici√≥n
- **Cu√°ndo usar**: An√°lisis de distribuciones

#### **üå°Ô∏è Mapas de Calor**
- **Prop√≥sito**: Mostrar intensidad en matriz
- **Ejemplo**: Rendimiento por equipo y mes
- **Cu√°ndo usar**: Datos categ√≥ricos bidimensionales

### 1.4 Errores Comunes en Visualizaci√≥n

#### **‚ùå Errores T√©cnicos**
- Ejes truncados que exageran diferencias
- Escalas inconsistentes entre gr√°ficos
- Uso inadecuado de gr√°ficos 3D

#### **‚ùå Errores de Dise√±o**
- Demasiados colores o patrones
- Texto demasiado peque√±o
- Leyendas confusas o innecesarias

#### **‚ùå Errores de Interpretaci√≥n**
- Confundir correlaci√≥n con causalidad
- Ignorar el contexto temporal
- Generalizar a partir de outliers

## 3. Datos de Pr√°ctica: Simulaci√≥n de Partidos

### 3.1 Creaci√≥n de Dataset Realista

Vamos a crear un dataset simulado que represente partidos de f√∫tbol con variables t√≠picas que encontrar√≠amos en an√°lisis reales:

- **Equipos**: Local vs. Visitante
- **Resultados**: Goles anotados por cada equipo
- **Contexto**: Liga, temporada, fecha
- **M√©tricas**: Posesi√≥n, tiros, faltas, tarjetas

````python
# Crear dataset simulado de partidos de f√∫tbol
np.random.seed(42)  # Para reproducibilidad

# Configuraci√≥n del dataset
n_partidos = 380  # T√≠pica temporada de liga
equipos = [
    'Real Madrid', 'Barcelona', 'Atl√©tico Madrid', 'Sevilla',
    'Real Sociedad', 'Betis', 'Villarreal', 'Athletic Bilbao',
    'Valencia', 'Osasuna', 'Celta', 'Rayo Vallecano',
    'Mallorca', 'Cadiz', 'Getafe', 'Espanyol',
    'Elche', 'Levante', 'Alaves', 'Granada'
]

# Generar datos de partidos
partidos_data = []

for i in range(n_partidos):
    # Seleccionar equipos
    equipos_partido = np.random.choice(equipos, 2, replace=False)
    local = equipos_partido[0]
    visitante = equipos_partido[1]
    
    # Ventaja de local (equipos locales tienden a anotar m√°s)
    ventaja_local = np.random.normal(0.3, 0.1)
    
    # Goles (siguiendo distribuci√≥n realista)
    goles_local = max(0, int(np.random.poisson(1.4 + ventaja_local)))
    goles_visitante = max(0, int(np.random.poisson(1.1)))
    
    # Resultado
    if goles_local > goles_visitante:
        resultado = 'Local'
    elif goles_visitante > goles_local:
        resultado = 'Visitante'
    else:
        resultado = 'Empate'
    
    # M√©tricas adicionales
    # Posesi√≥n (correlacionada con goles)
    posesion_local = np.random.normal(50 + (goles_local - goles_visitante) * 5, 10)
    posesion_local = np.clip(posesion_local, 20, 80)
    posesion_visitante = 100 - posesion_local
    
    # Tiros a porter√≠a
    tiros_local = max(1, int(np.random.poisson(4 + goles_local * 1.5)))
    tiros_visitante = max(1, int(np.random.poisson(4 + goles_visitante * 1.5)))
    
    # Faltas
    faltas_local = max(0, int(np.random.poisson(12)))
    faltas_visitante = max(0, int(np.random.poisson(12)))
    
    # Tarjetas
    tarjetas_local = max(0, int(np.random.poisson(2)))
    tarjetas_visitante = max(0, int(np.random.poisson(2)))
    
    # Fecha aleatoria en temporada
    fecha = pd.Timestamp('2023-08-01') + pd.Timedelta(days=np.random.randint(0, 280))
    
    partidos_data.append({
        'fecha': fecha,
        'local': local,
        'visitante': visitante,
        'goles_local': goles_local,
        'goles_visitante': goles_visitante,
        'resultado': resultado,
        'posesion_local': round(posesion_local, 1),
        'posesion_visitante': round(posesion_visitante, 1),
        'tiros_local': tiros_local,
        'tiros_visitante': tiros_visitante,
        'faltas_local': faltas_local,
        'faltas_visitante': faltas_visitante,
        'tarjetas_local': tarjetas_local,
        'tarjetas_visitante': tarjetas_visitante,
        'total_goles': goles_local + goles_visitante
    })

# Crear DataFrame
df_partidos = pd.DataFrame(partidos_data)

# Informaci√≥n del dataset
print("‚öΩ DATASET DE PARTIDOS GENERADO")
print("=" * 50)
print(f"üìä Total de partidos: {len(df_partidos)}")
print(f"üèüÔ∏è Equipos √∫nicos: {len(equipos)}")
print(f"üìÖ Per√≠odo: {df_partidos['fecha'].min().strftime('%Y-%m-%d')} a {df_partidos['fecha'].max().strftime('%Y-%m-%d')}")

# Estad√≠sticas b√°sicas
print(f"\n‚öΩ ESTAD√çSTICAS GENERALES:")
print(f"- Promedio goles por partido: {df_partidos['total_goles'].mean():.2f}")
print(f"- Victorias locales: {(df_partidos['resultado'] == 'Local').sum()} ({(df_partidos['resultado'] == 'Local').mean()*100:.1f}%)")
print(f"- Empates: {(df_partidos['resultado'] == 'Empate').sum()} ({(df_partidos['resultado'] == 'Empate').mean()*100:.1f}%)")
print(f"- Victorias visitantes: {(df_partidos['resultado'] == 'Visitante').sum()} ({(df_partidos['resultado'] == 'Visitante').mean()*100:.1f}%)")

print(f"\nüìã VISTA PREVIA DEL DATASET:")
print(df_partidos.head())
````

## 4. Pr√°ctica: Gr√°ficas Comparativas (Local vs. Visitante)

### 4.1 An√°lisis de Ventaja Local

Uno de los an√°lisis m√°s importantes en el f√∫tbol es determinar si existe realmente una "ventaja de local". Vamos a crear visualizaciones que nos ayuden a responder esta pregunta.

````python
# 4.1 An√°lisis de Ventaja Local vs. Visitante (Simplificado con Seaborn)
print("üèüÔ∏è AN√ÅLISIS DE VENTAJA LOCAL vs. VISITANTE")
print("=" * 50)

# Preparar datos para an√°lisis comparativo
datos_comparativos = []
for _, row in df_partidos.iterrows():
    datos_comparativos.append({
        'equipo': 'Local',
        'goles': row['goles_local'],
        'posesion': row['posesion_local'],
        'tiros': row['tiros_local']
    })
    datos_comparativos.append({
        'equipo': 'Visitante', 
        'goles': row['goles_visitante'],
        'posesion': row['posesion_visitante'],
        'tiros': row['tiros_visitante']
    })

df_comparativo = pd.DataFrame(datos_comparativos)

# Crear visualizaciones con seaborn (m√°s simple)
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('üèüÔ∏è AN√ÅLISIS DE VENTAJA LOCAL vs. VISITANTE', fontsize=16, fontweight='bold')

# Gr√°fico 1: Goles promedio (barplot)
sns.barplot(data=df_comparativo, x='equipo', y='goles', ax=axes[0, 0])
axes[0, 0].set_title('Goles Promedio por Partido', fontweight='bold')

# Gr√°fico 2: Distribuci√≥n de resultados
resultado_counts = df_partidos['resultado'].value_counts()
axes[0, 1].pie(resultado_counts.values, labels=resultado_counts.index, autopct='%1.1f%%', startangle=90)
axes[0, 1].set_title('Distribuci√≥n de Resultados', fontweight='bold')

# Gr√°fico 3: Posesi√≥n promedio
sns.barplot(data=df_comparativo, x='equipo', y='posesion', ax=axes[1, 0])
axes[1, 0].set_title('Posesi√≥n Promedio (%)', fontweight='bold')

# Gr√°fico 4: Tiros a porter√≠a
sns.barplot(data=df_comparativo, x='equipo', y='tiros', ax=axes[1, 1])
axes[1, 1].set_title('Tiros a Porter√≠a Promedio', fontweight='bold')

plt.tight_layout()
plt.show()

# An√°lisis estad√≠stico simplificado
print("\nüìä RESUMEN ESTAD√çSTICO:")
resumen = df_comparativo.groupby('equipo').agg({
    'goles': 'mean',
    'posesion': 'mean', 
    'tiros': 'mean'
}).round(2)

print(resumen)

# Conclusi√≥n
diferencia_goles = resumen.loc['Local', 'goles'] - resumen.loc['Visitante', 'goles']
print(f"\nüí° CONCLUSI√ìN: {'S√ç' if diferencia_goles > 0 else 'NO'} existe ventaja local")
print(f"üìà Diferencia de goles: {diferencia_goles:.2f}")
````

## 5. Visualizaciones Avanzadas con Seaborn

### 5.1 Distribuciones y Patrones

Seaborn nos permite crear visualizaciones m√°s sofisticadas y estad√≠sticamente informativas. Vamos a explorar las distribuciones de goles y buscar patrones interesantes.

````python
# 5.1 Visualizaciones Avanzadas con Seaborn (Simplificado)
print("üìä AN√ÅLISIS AVANZADO CON SEABORN")
print("=" * 50)

# Preparar datos en formato largo para seaborn
datos_goles = []
for _, row in df_partidos.iterrows():
    datos_goles.append({'condicion': 'Local', 'goles': row['goles_local']})
    datos_goles.append({'condicion': 'Visitante', 'goles': row['goles_visitante']})

df_goles = pd.DataFrame(datos_goles)

# Preparar datos de correlaci√≥n
datos_metricas = []
for _, row in df_partidos.iterrows():
    datos_metricas.extend([
        {'goles': row['goles_local'], 'posesion': row['posesion_local'], 
         'tiros': row['tiros_local'], 'faltas': row['faltas_local']},
        {'goles': row['goles_visitante'], 'posesion': row['posesion_visitante'],
         'tiros': row['tiros_visitante'], 'faltas': row['faltas_visitante']}
    ])

df_metricas = pd.DataFrame(datos_metricas)

# Crear visualizaciones con seaborn
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('üìä AN√ÅLISIS AVANZADO CON SEABORN', fontsize=16, fontweight='bold')

# 1. Distribuci√≥n de goles (violin plot)
sns.violinplot(data=df_goles, x='condicion', y='goles', ax=axes[0, 0])
axes[0, 0].set_title('Distribuci√≥n de Goles', fontweight='bold')

# 2. Relaci√≥n posesi√≥n vs goles
sns.scatterplot(data=df_metricas, x='posesion', y='goles', ax=axes[0, 1], alpha=0.6)
sns.regplot(data=df_metricas, x='posesion', y='goles', ax=axes[0, 1], scatter=False, color='red')
axes[0, 1].set_title('Posesi√≥n vs. Goles', fontweight='bold')

# 3. Matriz de correlaciones (heatmap)
correlacion = df_metricas.corr()
sns.heatmap(correlacion, annot=True, cmap='coolwarm', center=0, ax=axes[1, 0])
axes[1, 0].set_title('Correlaciones', fontweight='bold')

# 4. Goles por mes (boxplot)
df_partidos['mes'] = df_partidos['fecha'].dt.strftime('%b')
sns.boxplot(data=df_partidos, x='mes', y='total_goles', ax=axes[1, 1])
axes[1, 1].set_title('Goles por Mes', fontweight='bold')
axes[1, 1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# An√°lisis de correlaciones (simplificado)
print("\nüîç CORRELACIONES PRINCIPALES:")
print(f"üìà Posesi√≥n-Goles: {correlacion.loc['posesion', 'goles']:.3f}")
print(f"üéØ Tiros-Goles: {correlacion.loc['tiros', 'goles']:.3f}")
print(f"‚ö†Ô∏è Faltas-Goles: {correlacion.loc['faltas', 'goles']:.3f}")

# Insights autom√°ticos
if correlacion.loc['posesion', 'goles'] > 0.3:
    print("\nüí° INSIGHT: Mayor posesi√≥n tiende a generar m√°s goles")
if correlacion.loc['tiros', 'goles'] > 0.5:
    print("üí° INSIGHT: Los tiros est√°n fuertemente correlacionados con los goles")
````

## 6. Replicando "La Remontada": An√°lisis Temporal

### 6.1 Simulaci√≥n de un Partido √âpico

Vamos a replicar el tipo de an√°lisis que se har√≠a para un partido legendario como "La Remontada" del PSG vs. Barcelona (6-1, 2017). Crearemos visualizaciones que muestren la evoluci√≥n temporal de un partido.

````python
# 6.1 Simulaci√≥n de "La Remontada" - An√°lisis Temporal Simplificado
np.random.seed(95)  # Para crear una "remontada" espec√≠fica

# Eventos de gol predefinidos para la narrativa
eventos_gol = {
    12: ('visitante', 'Gol temprano del visitante'),
    25: ('visitante', 'Segundo gol visitante'), 
    40: ('visitante', 'Tercer gol visitante'),
    50: ('local', 'Primer gol local - inicia la remontada'),
    65: ('local', 'Segundo gol local'),
    75: ('local', 'Tercer gol local - empate √©pico'),
    85: ('local', 'Cuarto gol local - remontada completada')
}

# Simular evoluci√≥n del partido
minutos = list(range(1, 91))
goles_local = goles_visitante = 0
datos_temporal = []

for minuto in minutos:
    # Verificar si hay gol
    if minuto in eventos_gol:
        equipo_gol, _ = eventos_gol[minuto]
        if equipo_gol == 'local':
            goles_local += 1
        else:
            goles_visitante += 1
    
    # Simular posesi√≥n seg√∫n marcador
    if goles_local > goles_visitante:
        posesion = np.random.normal(60, 8)
    elif goles_visitante > goles_local:
        posesion = np.random.normal(40, 8)
    else:
        posesion = np.random.normal(50, 5)
    
    posesion = np.clip(posesion, 25, 75)
    
    datos_temporal.append({
        'minuto': minuto,
        'goles_local': goles_local,
        'goles_visitante': goles_visitante,
        'posesion_local': posesion
    })

df_temporal = pd.DataFrame(datos_temporal)

# Crear visualizaciones simplificadas con seaborn
fig, axes = plt.subplots(2, 1, figsize=(14, 10))
fig.suptitle('‚öΩ "LA REMONTADA" - AN√ÅLISIS TEMPORAL', fontsize=16, fontweight='bold')

# Evoluci√≥n del marcador
sns.lineplot(data=df_temporal, x='minuto', y='goles_local', 
             marker='o', linewidth=3, label='Equipo Local', ax=axes[0])
sns.lineplot(data=df_temporal, x='minuto', y='goles_visitante',
             marker='s', linewidth=3, label='Equipo Visitante', ax=axes[0])
axes[0].set_title('Evoluci√≥n del Marcador', fontweight='bold')
axes[0].set_ylabel('Goles Acumulados')

# Evoluci√≥n de la posesi√≥n
sns.lineplot(data=df_temporal, x='minuto', y='posesion_local', 
             linewidth=2, ax=axes[1], color='green')
axes[1].axhline(y=50, color='red', linestyle='--', label='Equilibrio (50%)')
axes[1].fill_between(df_temporal['minuto'], df_temporal['posesion_local'], 50, alpha=0.3)
axes[1].set_title('Evoluci√≥n de la Posesi√≥n Local', fontweight='bold')
axes[1].set_ylabel('Posesi√≥n Local (%)')
axes[1].set_xlabel('Minuto')

plt.tight_layout()
plt.show()

# Estad√≠sticas simplificadas
print("üìä ESTAD√çSTICAS DEL PARTIDO √âPICO")
print("=" * 50)
print(f"üèÜ Resultado final: {goles_local} - {goles_visitante}")
print(f"üìà Remontada hist√≥rica: De 0-3 a {goles_local}-3")
print(f"‚è±Ô∏è Posesi√≥n promedio local: {df_temporal['posesion_local'].mean():.1f}%")

# Momentos clave
print(f"\nüéØ MOMENTOS CLAVE:")
for minuto, (equipo, descripcion) in eventos_gol.items():
    print(f"‚öΩ Minuto {minuto}: {descripcion}")
````
