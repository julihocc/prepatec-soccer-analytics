## 1. El Escenario: Presentando un Nuevo Modelo de Scouting

Imagina que has desarrollado un modelo que predice el potencial de jugadores jóvenes. Ahora debes convencer al club de que adopte tu modelo. Prepararás:

1.  **Una presentación para la Junta Directiva** (enfocada en el impacto y el ROI).

## 2. Estructura de la Presentación para la Junta Directiva

**Objetivo:** Conseguir aprobación y financiamiento.
**Tono:** Ejecutivo, centrado en el negocio.
**Duración:** 10-15 minutos.

### Diapositivas Clave:

1.  **Título:** "Revolucionando el Scouting: Un Enfoque Basado en Datos para Identificar Talento".
2.  **El Problema:** "El mercado de fichajes es ineficiente y costoso. Estamos perdiendo talento infravalorado". (Mostrar 1-2 estadísticas de mercado impactantes).
3.  **La Solución:** "Hemos desarrollado un modelo predictivo que identifica jugadores con alto potencial y bajo costo". (Visualización simple del concepto).
4.  **¿Cómo Funciona? (Simplificado):** "Analizamos más de 50 métricas de rendimiento, físicas y de juego para generar un 'Score de Potencial'". (Diagrama de alto nivel, sin jerga técnica).
5.  **Validación del Modelo:** "Nuestro modelo supera en un 30% la precisión de los métodos actuales en la identificación de futuros jugadores de élite". (Gráfico de barras comparativo muy claro).
6.  **Caso de Éxito (Ejemplo):** "El modelo identificó a 'Jugador X', fichado por €1M, cuyo valor actual se proyecta en €15M". (Foto del jugador, métricas clave de antes y después).
7.  **Impacto en el Negocio (ROI):** "Proyectamos un ahorro de €5M en fichajes y un aumento de €20M en valor de plantilla en 3 años". (Gráfico de proyecciones financieras).
8.  **Próximos Pasos y Solicitud:** "Solicitamos €150,000 para integrar el modelo en nuestras plataformas de scouting y contratar un analista de datos". (Línea de tiempo simple).
9.  **El Equipo:** Presentar brevemente al equipo detrás del proyecto.
10. **Preguntas y Cierre:** Resumen de los beneficios clave.

## 3. Estructura de la Presentación para el Equipo de Scouting

**Objetivo:** Lograr que los scouts confíen y usen la herramienta.
**Tono:** Colaborativo, práctico, técnico pero accesible.
**Duración:** 30-45 minutos (con demo).

### Diapositivas Clave:

1.  **Título:** "Potenciando la Intuición del Scout con Datos: Una Nueva Herramienta de Soporte".
2.  **Introducción:** "Esta herramienta está diseñada para complementar vuestra experiencia, no para reemplazarla".
3.  **¿Qué problema resolvemos juntos?** "Cómo podemos cubrir más ligas y analizar más jugadores de manera eficiente".
4.  **Metodología (¿Qué hay dentro de la 'caja negra'?):**
    -   **Fuentes de Datos:** Listar las fuentes de datos utilizadas (StatsBomb, WyScout, etc.).
    -   **Características Principales:** Explicar qué métricas se usan (xG, pases progresivos, duelos aéreos, etc.) y por qué son importantes.
    -   **El Modelo:** Explicar brevemente que es un `Random Forest` y por qué se eligió (robusto, interpretable).
5.  **Interpretación de Resultados:** "El 'Score de Potencial' se compone de tres sub-scores: Técnico, Táctico y Físico". (Mostrar cómo se desglosa el score).
6.  **DEMO EN VIVO:** Mostrar el dashboard interactivo. Buscar un jugador, analizar su perfil, encontrar jugadores similares.
7.  **Validación y Limitaciones:**
    -   **Casos de Éxito:** Mostrar jugadores que el modelo valoró correctamente.
    -   **Casos de 'Fallo':** Mostrar jugadores donde el modelo se equivocó y discutir por qué (lesiones, factores psicológicos). Ser transparente para generar confianza.
8.  **Integración en vuestro Flujo de Trabajo:** "El modelo genera un reporte semanal con 10 jugadores recomendados para que los analicéis en video".
9.  **Feedback y Próximos Pasos:** "Queremos vuestro feedback para mejorar el modelo. ¿Qué métricas faltan? ¿Qué os parece la interfaz?".
10. **Taller Práctico:** Sesión de preguntas y respuestas donde los scouts pueden usar la herramienta.

## 4. Ejercicio Práctico: Crear Visualizaciones Clave

Usando `plotly` o `matplotlib`, crea las siguientes visualizaciones para tus presentaciones.

```python
import plotly.graph_objects as go
import pandas as pd

# Gráfico para la Junta Directiva: ROI
fig_roi = go.Figure([
    go.Bar(name='Ahorro en Fichajes (€M)', x=['Año 1', 'Año 2', 'Año 3'], y=[1, 3, 5]),
    go.Bar(name='Aumento Valor Plantilla (€M)', x=['Año 1', 'Año 2', 'Año 3'], y=[5, 12, 20])
])
fig_roi.update_layout(
    barmode='group',
    title='Proyección de Impacto Financiero a 3 Años',
    yaxis_title='Millones de Euros (€)'
)
fig_roi.show()

# Gráfico para Scouts: Comparación de Jugadores (Radar Chart)
categories = ['Ataque', 'Defensa', 'Técnica', 'Físico', 'Mentalidad']

fig_radar = go.Figure()

fig_radar.add_trace(go.Scatterpolar(
      r=[85, 60, 90, 75, 80],
      theta=categories,
      fill='toself',
      name='Jugador Recomendado'
))
fig_radar.add_trace(go.Scatterpolar(
      r=[70, 75, 75, 85, 88],
      theta=categories,
      fill='toself',
      name='Jugador Promedio (Misma Posición)'
))

fig_radar.update_layout(
  polar=dict(
    radialaxis=dict(
      visible=True,
      range=[0, 100]
    )),
  showlegend=True,
  title="Perfil Comparativo del Jugador"
)

fig_radar.show()
```
