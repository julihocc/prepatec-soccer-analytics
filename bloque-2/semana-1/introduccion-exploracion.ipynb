## 2. Preparando Nuestras Herramientas

Necesitamos importar las herramientas para trabajar con datos de f√∫tbol.

**Lo que vamos a usar:**
- üìä **pandas**: Para leer datos como si fuera Excel

```python
# Importar las herramientas principales
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Configurar seaborn para gr√°ficos bonitos
sns.set_theme(style="whitegrid", palette="viridis")

print("‚úÖ ¬°Herramientas listas!")
print("üìä Pandas: Para leer datos de f√∫tbol")
print("üé® Seaborn: Para gr√°ficos incre√≠bles")
print("‚öΩ ¬°Listos para explorar datos deportivos!")
```

### ¬øQu√© es Pandas?

Pandas es como Excel, pero m√°s poderoso. Nos ayuda a:
- üìñ Leer datos de archivos
- üîç Buscar informaci√≥n espec√≠fica
- üßÆ Hacer c√°lculos autom√°ticamente

**Ejemplo simple:** Una tabla con equipos y sus goles

```python
# Una lista simple de equipos favoritos
equipos_favoritos = ['Barcelona', 'Real Madrid', 'Manchester City', 'Bayern Munich']

print("‚öΩ Mis equipos favoritos:")
for i, equipo in enumerate(equipos_favoritos, 1):
    print(f"{i}. {equipo}")

print(f"\nüî¢ Total de equipos: {len(equipos_favoritos)}")
```

```python
# Crear una tabla simple con datos de equipos
datos_equipos = {
    'Equipo': ['Barcelona', 'Real Madrid', 'Manchester City'],
    'Goles': [68, 87, 95],
    'Partidos': [38, 38, 38]
}

tabla_equipos = pd.DataFrame(datos_equipos)
print("üìä Tabla de equipos:")
print(tabla_equipos)

print(f"\n‚ú® ¬°Esta tabla tiene {len(tabla_equipos)} equipos y {len(tabla_equipos.columns)} columnas!")
```

## 3. ¬°Datos Reales de F√∫tbol!

Ahora vamos a trabajar con datos reales de partidos.

**¬øQu√© contienen estos datos?**
- üèüÔ∏è Equipos que jugaron
- ‚öΩ Goles marcados
- üìÖ Cu√°ndo se jug√≥
- üèÜ En qu√© competici√≥n

¬°Vamos a crear datos de ejemplo que parecen reales!

```python
# Crear datos simples de partidos de f√∫tbol
import random

equipos = ['Barcelona', 'Real Madrid', 'Manchester City', 'Bayern Munich', 
          'Liverpool', 'Paris Saint-Germain', 'Chelsea', 'Juventus']

# Crear lista de partidos
partidos = []
for i in range(50):  # 50 partidos
    equipo_local = random.choice(equipos)
    equipo_visitante = random.choice([e for e in equipos if e != equipo_local])
    
    partidos.append({
        'Equipo_Local': equipo_local,
        'Equipo_Visitante': equipo_visitante,
        'Goles_Local': random.randint(0, 4),
        'Goles_Visitante': random.randint(0, 3),
        'Temporada': random.choice(['2022-23', '2023-24'])
    })

# Convertir a tabla
datos_futbol = pd.DataFrame(partidos)
print("‚öΩ ¬°Datos de f√∫tbol creados!")
print(f"üìä Tenemos {len(datos_futbol)} partidos")
print(f"üèüÔ∏è Con {len(equipos)} equipos diferentes")
```

### Echemos un Vistazo a Nuestros Datos

¬°Vamos a ver qu√© partidos tenemos!

```python
# Ver los primeros 10 partidos
print("üîç Los primeros 10 partidos:")
print(datos_futbol.head(10))

print(f"\nüìà Resumen r√°pido:")
print(f"‚Ä¢ Total de partidos: {len(datos_futbol)}")
print(f"‚Ä¢ Columnas de informaci√≥n: {list(datos_futbol.columns)}")
```

```python
# Calcular algunas estad√≠sticas interesantes
total_goles = datos_futbol['Goles_Local'].sum() + datos_futbol['Goles_Visitante'].sum()
promedio_goles = total_goles / len(datos_futbol)

print("‚öΩ Estad√≠sticas B√°sicas:")
print(f"‚Ä¢ Total de goles en todos los partidos: {total_goles}")
print(f"‚Ä¢ Promedio de goles por partido: {promedio_goles:.1f}")
print(f"‚Ä¢ Gol m√°ximo en un partido (local): {datos_futbol['Goles_Local'].max()}")
print(f"‚Ä¢ Gol m√°ximo en un partido (visitante): {datos_futbol['Goles_Visitante'].max()}")
```

## 4. ¬°Hagamos Gr√°ficos Bonitos con Seaborn!

Ahora viene lo divertido: convertir n√∫meros en gr√°ficos incre√≠bles.

```python
# Gr√°fico 1: ¬øCu√°ntos goles marcan los equipos locales?
plt.figure(figsize=(10, 6))
sns.countplot(data=datos_futbol, x='Goles_Local', palette='viridis')
plt.title('¬øCu√°ntos Goles Marcan los Equipos Locales?', fontsize=16, fontweight='bold')
plt.xlabel('Goles Marcados')
plt.ylabel('N√∫mero de Partidos')
plt.show()

print("üìä ¬°Este gr√°fico nos muestra qu√© tan seguido los equipos marcan 0, 1, 2, 3 o 4 goles en casa!")
```

```python
# Gr√°fico 2: ¬øEs mejor jugar en casa?
datos_futbol['Total_Goles'] = datos_futbol['Goles_Local'] + datos_futbol['Goles_Visitante']

plt.figure(figsize=(10, 6))
sns.boxplot(data=datos_futbol, y='Total_Goles', palette='Set2')
plt.title('¬øCu√°ntos Goles Hay por Partido?', fontsize=16, fontweight='bold')
plt.ylabel('Total de Goles en el Partido')
plt.show()

print("üìà ¬°Este gr√°fico nos muestra si hay partidos con muchos o pocos goles!")
print(f"üéØ Promedio de goles por partido: {datos_futbol['Total_Goles'].mean():.1f}")
```

```python
# Gr√°fico 3: ¬øQu√© equipos juegan m√°s partidos?
equipos_conteo = datos_futbol['Equipo_Local'].value_counts()

plt.figure(figsize=(12, 6))
sns.barplot(x=equipos_conteo.values, y=equipos_conteo.index, palette='plasma')
plt.title('¬øQu√© Equipos Aparecen M√°s Como Locales?', fontsize=16, fontweight='bold')
plt.xlabel('N√∫mero de Partidos como Local')
plt.ylabel('Equipos')
plt.show()

print("üèüÔ∏è ¬°Este gr√°fico muestra qu√© equipos jugaron m√°s partidos en casa!")
print(f"üéØ El equipo que m√°s jug√≥ en casa: {equipos_conteo.index[0]} con {equipos_conteo.iloc[0]} partidos")
```

```python
# ¬øEs mejor jugar en casa? ¬°Vamos a descubrirlo!
# Crear una nueva columna que diga qui√©n gan√≥
def quien_gano(fila):
    if fila['Goles_Local'] > fila['Goles_Visitante']:
        return 'Gan√≥ Local'
    elif fila['Goles_Local'] < fila['Goles_Visitante']:
        return 'Gan√≥ Visitante'
    else:
        return 'Empate'

datos_futbol['Resultado'] = datos_futbol.apply(quien_gano, axis=1)

# Hacer gr√°fico con seaborn
plt.figure(figsize=(10, 6))
sns.countplot(data=datos_futbol, x='Resultado', palette='Set1')
plt.title('¬øEs Mejor Jugar en Casa?', fontsize=16, fontweight='bold')
plt.xlabel('Resultado del Partido')
plt.ylabel('N√∫mero de Partidos')
plt.show()

# Mostrar porcentajes
resultados = datos_futbol['Resultado'].value_counts()
total = len(datos_futbol)
print("üèüÔ∏è Ventaja de jugar en casa:")
for resultado, cantidad in resultados.items():
    porcentaje = (cantidad / total) * 100
    print(f"‚Ä¢ {resultado}: {cantidad} partidos ({porcentaje:.1f}%)")
```

## 5. ¬°Tu Turno! Ejercicio Pr√°ctico

Ahora vamos a hacer algunos descubrimientos por nuestra cuenta.

```python
# Ejercicio 1: ¬°Encuentra el partido con m√°s goles!
datos_futbol['Total_Goles'] = datos_futbol['Goles_Local'] + datos_futbol['Goles_Visitante']
partido_mas_goles = datos_futbol.loc[datos_futbol['Total_Goles'].idxmax()]

print("üéØ El partido con m√°s goles fue:")
print(f"‚öΩ {partido_mas_goles['Equipo_Local']} {partido_mas_goles['Goles_Local']} - {partido_mas_goles['Goles_Visitante']} {partido_mas_goles['Equipo_Visitante']}")
print(f"‚öΩ Total de goles: {partido_mas_goles['Total_Goles']}")

# Ejercicio 2: ¬øCu√°l es tu equipo favorito?
mi_equipo_favorito = "Barcelona"  # ¬°Cambia esto por tu equipo favorito!

partidos_favorito = datos_futbol[
    (datos_futbol['Equipo_Local'] == mi_equipo_favorito) | 
    (datos_futbol['Equipo_Visitante'] == mi_equipo_favorito)
]

print(f"\n‚öΩ Partidos de {mi_equipo_favorito}:")
print(f"üéÆ Total de partidos: {len(partidos_favorito)}")

if len(partidos_favorito) > 0:
    print("üìã Algunos partidos:")
    print(partidos_favorito[['Equipo_Local', 'Goles_Local', 'Goles_Visitante', 'Equipo_Visitante']].head())
else:
    print("üòî Tu equipo no est√° en estos datos. ¬°Prueba con otro!")
```

```python
# An√°lisis m√°s profundo de nuestros datos de f√∫tbol
# Vamos a crear visualizaciones m√°s interesantes

# Crear nuevas columnas para an√°lisis
datos_futbol['diferencia_goles'] = datos_futbol['Goles_Local'] - datos_futbol['Goles_Visitante']

# Crear visualizaciones con subplots
fig, axes = plt.subplots(2, 2, figsize=(15, 10))

# 1. Distribuci√≥n de goles por temporada
sns.boxplot(data=datos_futbol, x='Temporada', y='Total_Goles', ax=axes[0,0], palette='Set2')
axes[0,0].set_title('Distribuci√≥n de Goles por Temporada', fontsize=12, fontweight='bold')
axes[0,0].set_xlabel('Temporada')
axes[0,0].set_ylabel('Total de Goles')

# 2. Goles locales vs visitantes
sns.scatterplot(data=datos_futbol, x='Goles_Local', y='Goles_Visitante', 
                hue='Resultado', ax=axes[0,1], s=80, alpha=0.7)
axes[0,1].set_title('Goles Locales vs Visitantes', fontsize=12, fontweight='bold')
axes[0,1].set_xlabel('Goles Equipo Local')
axes[0,1].set_ylabel('Goles Equipo Visitante')

# 3. Diferencia de goles por resultado
sns.violinplot(data=datos_futbol, x='Resultado', y='diferencia_goles', 
               ax=axes[1,0], palette='viridis')
axes[1,0].set_title('Diferencia de Goles por Resultado', fontsize=12, fontweight='bold')
axes[1,0].set_xlabel('Resultado del Partido')
axes[1,0].set_ylabel('Diferencia de Goles (Local - Visitante)')

# 4. Distribuci√≥n de goles totales
sns.histplot(data=datos_futbol, x='Total_Goles', bins=8, 
             ax=axes[1,1], color='skyblue', alpha=0.7)
axes[1,1].axvline(datos_futbol['Total_Goles'].mean(), color='red', 
                  linestyle='--', label=f'Promedio: {datos_futbol["Total_Goles"].mean():.1f}')
axes[1,1].set_title('Distribuci√≥n de Goles Totales por Partido', fontsize=12, fontweight='bold')
axes[1,1].set_xlabel('Total de Goles')
axes[1,1].set_ylabel('Frecuencia')
axes[1,1].legend()

plt.tight_layout()
plt.show()

print("üé® ¬°An√°lisis visual completado!")
print("üìä Estas gr√°ficas muestran patrones interesantes:")
print(f"‚öΩ Promedio de goles por partido: {datos_futbol['Total_Goles'].mean():.1f}")
print(f"üèüÔ∏è Diferencia promedio (local-visitante): {datos_futbol['diferencia_goles'].mean():.1f}")
print(f"üìà Rango de goles: {datos_futbol['Total_Goles'].min()} - {datos_futbol['Total_Goles'].max()}")
```

```python
# Verificaci√≥n final de nuestros datos de f√∫tbol
print("üîç Informaci√≥n completa de nuestros datos:")
print(f"üìä Total de filas: {len(datos_futbol)}")
print(f"üìã Total de columnas: {len(datos_futbol.columns)}")

# Verificar datos faltantes
print("\nüîç Verificaci√≥n de datos faltantes:")
print(datos_futbol.isnull().sum())

# Mostrar tipos de datos
print("\nüìã Tipos de datos:")
print(datos_futbol.dtypes)

# Estad√≠sticas por temporada
print("\n‚öΩ Estad√≠sticas por temporada:")
stats_temporada = datos_futbol.groupby('Temporada').agg({
    'Total_Goles': ['mean', 'sum', 'count'],
    'Goles_Local': 'mean',
    'Goles_Visitante': 'mean'
}).round(2)

for temporada in datos_futbol['Temporada'].unique():
    temp_data = datos_futbol[datos_futbol['Temporada'] == temporada]
    print(f"\nüèÜ Temporada {temporada}:")
    print(f"  ‚Ä¢ Partidos jugados: {len(temp_data)}")
    print(f"  ‚Ä¢ Promedio goles/partido: {temp_data['Total_Goles'].mean():.1f}")
    print(f"  ‚Ä¢ Total goles temporada: {temp_data['Total_Goles'].sum()}")
    
# Verificar valores √∫nicos en columnas importantes
print("\nüî¢ Valores √∫nicos por columna:")
for col in ['Temporada', 'Resultado']:
    if col in datos_futbol.columns:
        valores_unicos = datos_futbol[col].value_counts()
        print(f"\n{col}:")
        for valor, cantidad in valores_unicos.items():
            print(f"  ‚Ä¢ {valor}: {cantidad} registros")

print("\n‚úÖ ¬°Verificaci√≥n de datos completada!")
```
