## 1. Teor√≠a: Principios de Visualizaci√≥n Efectiva

### 1.1 ¬øPor qu√© Visualizar Datos?

La visualizaci√≥n de datos es **fundamental** en el an√°lisis deportivo porque:

#### **üß† Percepci√≥n Visual**
- El cerebro humano procesa informaci√≥n visual **mucho m√°s r√°pido** que texto
- Los patrones son m√°s f√°ciles de identificar en gr√°ficos que en tablas
- La memoria visual es m√°s duradera que la textual

#### **üìä Comunicaci√≥n**
- **Simplifica** conceptos complejos
- **Persuade** con evidencia visual
- **Unifica** la comprensi√≥n entre audiencias diversas

#### **‚öΩ An√°lisis Deportivo**
- **Identifica tendencias** en rendimiento
- **Compara** jugadores y equipos de forma intuitiva
- **Detecta anomal√≠as** y oportunidades

### 1.2 Principios de Visualizaci√≥n Efectiva

#### **1.2.1 Claridad y Simplicidad**
- **Una historia por gr√°fico**: Enfoque en un mensaje principal
- **Eliminar ruido visual**: Menos elementos = mayor impacto
- **Jerarqu√≠a visual**: Destacar lo m√°s importante

#### **1.2.2 Precisi√≥n e Integridad**
- **Escalas apropiadas**: Empezar ejes en cero cuando sea relevante
- **Proporciones reales**: Evitar distorsiones enga√±osas
- **Contexto completo**: Incluir informaci√≥n necesaria para interpretaci√≥n

#### **1.2.3 Est√©tica y Profesionalismo**
- **Colores significativos**: Paletas que refuercen el mensaje
- **Tipograf√≠a legible**: Tama√±os y fuentes apropiadas
- **Consistencia**: Mismo estilo en toda la presentaci√≥n

### 1.3 Tipos de Gr√°ficos y Cu√°ndo Usarlos

#### **üìà Gr√°ficos de L√≠neas**
- **Prop√≥sito**: Mostrar cambios en el tiempo
- **Ejemplo**: Evoluci√≥n de goles por temporada
- **Cu√°ndo usar**: Datos temporales, tendencias

#### **üìä Gr√°ficos de Barras**
- **Prop√≥sito**: Comparar categor√≠as
- **Ejemplo**: Goles por equipo en una temporada
- **Cu√°ndo usar**: Comparaciones entre grupos

#### **üîç Gr√°ficos de Dispersi√≥n**
- **Prop√≥sito**: Mostrar relaciones entre variables
- **Ejemplo**: Posesi√≥n vs. goles anotados
- **Cu√°ndo usar**: Correlaciones, patrones

#### **üì¶ Boxplots**
- **Prop√≥sito**: Mostrar distribuci√≥n y outliers
- **Ejemplo**: Distribuci√≥n de edades por posici√≥n
- **Cu√°ndo usar**: An√°lisis de distribuciones

#### **üå°Ô∏è Mapas de Calor**
- **Prop√≥sito**: Mostrar intensidad en matriz
- **Ejemplo**: Rendimiento por equipo y mes
- **Cu√°ndo usar**: Datos categ√≥ricos bidimensionales

### 1.4 Errores Comunes en Visualizaci√≥n

#### **‚ùå Errores T√©cnicos**
- Ejes truncados que exageran diferencias
- Escalas inconsistentes entre gr√°ficos
- Uso inadecuado de gr√°ficos 3D

#### **‚ùå Errores de Dise√±o**
- Demasiados colores o patrones
- Texto demasiado peque√±o
- Leyendas confusas o innecesarias

#### **‚ùå Errores de Interpretaci√≥n**
- Confundir correlaci√≥n con causalidad
- Ignorar el contexto temporal
- Generalizar a partir de outliers

## 3. Datos de Pr√°ctica: Simulaci√≥n de Partidos

### 3.1 Creaci√≥n de Dataset Realista

Vamos a crear un dataset simulado que represente partidos de f√∫tbol con variables t√≠picas que encontrar√≠amos en an√°lisis reales:

- **Equipos**: Local vs. Visitante
- **Resultados**: Goles anotados por cada equipo
- **Contexto**: Liga, temporada, fecha
- **M√©tricas**: Posesi√≥n, tiros, faltas, tarjetas

```python
# Crear dataset simulado de partidos de f√∫tbol
np.random.seed(42)  # Para reproducibilidad

# Configuraci√≥n del dataset
n_partidos = 380  # T√≠pica temporada de liga
equipos = [
    'Real Madrid', 'Barcelona', 'Atl√©tico Madrid', 'Sevilla',
    'Real Sociedad', 'Betis', 'Villarreal', 'Athletic Bilbao',
    'Valencia', 'Osasuna', 'Celta', 'Rayo Vallecano',
    'Mallorca', 'Cadiz', 'Getafe', 'Espanyol',
    'Elche', 'Levante', 'Alaves', 'Granada'
]

# Generar datos de partidos
partidos_data = []

for i in range(n_partidos):
    # Seleccionar equipos
    equipos_partido = np.random.choice(equipos, 2, replace=False)
    local = equipos_partido[0]
    visitante = equipos_partido[1]
    
    # Ventaja de local (equipos locales tienden a anotar m√°s)
    ventaja_local = np.random.normal(0.3, 0.1)
    
    # Goles (siguiendo distribuci√≥n realista)
    goles_local = max(0, int(np.random.poisson(1.4 + ventaja_local)))
    goles_visitante = max(0, int(np.random.poisson(1.1)))
    
    # Resultado
    if goles_local > goles_visitante:
        resultado = 'Local'
    elif goles_visitante > goles_local:
        resultado = 'Visitante'
    else:
        resultado = 'Empate'
    
    # M√©tricas adicionales
    # Posesi√≥n (correlacionada con goles)
    posesion_local = np.random.normal(50 + (goles_local - goles_visitante) * 5, 10)
    posesion_local = np.clip(posesion_local, 20, 80)
    posesion_visitante = 100 - posesion_local
    
    # Tiros a porter√≠a
    tiros_local = max(1, int(np.random.poisson(4 + goles_local * 1.5)))
    tiros_visitante = max(1, int(np.random.poisson(4 + goles_visitante * 1.5)))
    
    # Faltas
    faltas_local = max(0, int(np.random.poisson(12)))
    faltas_visitante = max(0, int(np.random.poisson(12)))
    
    # Tarjetas
    tarjetas_local = max(0, int(np.random.poisson(2)))
    tarjetas_visitante = max(0, int(np.random.poisson(2)))
    
    # Fecha aleatoria en temporada
    fecha = pd.Timestamp('2023-08-01') + pd.Timedelta(days=np.random.randint(0, 280))
    
    partidos_data.append({
        'fecha': fecha,
        'local': local,
        'visitante': visitante,
        'goles_local': goles_local,
        'goles_visitante': goles_visitante,
        'resultado': resultado,
        'posesion_local': round(posesion_local, 1),
        'posesion_visitante': round(posesion_visitante, 1),
        'tiros_local': tiros_local,
        'tiros_visitante': tiros_visitante,
        'faltas_local': faltas_local,
        'faltas_visitante': faltas_visitante,
        'tarjetas_local': tarjetas_local,
        'tarjetas_visitante': tarjetas_visitante,
        'total_goles': goles_local + goles_visitante
    })

# Crear DataFrame
df_partidos = pd.DataFrame(partidos_data)

# Informaci√≥n del dataset
print("‚öΩ DATASET DE PARTIDOS GENERADO")
print("=" * 50)
print(f"üìä Total de partidos: {len(df_partidos)}")
print(f"üèüÔ∏è Equipos √∫nicos: {len(equipos)}")
print(f"üìÖ Per√≠odo: {df_partidos['fecha'].min().strftime('%Y-%m-%d')} a {df_partidos['fecha'].max().strftime('%Y-%m-%d')}")

# Estad√≠sticas b√°sicas
print(f"\n‚öΩ ESTAD√çSTICAS GENERALES:")
print(f"- Promedio goles por partido: {df_partidos['total_goles'].mean():.2f}")
print(f"- Victorias locales: {(df_partidos['resultado'] == 'Local').sum()} ({(df_partidos['resultado'] == 'Local').mean()*100:.1f}%)")
print(f"- Empates: {(df_partidos['resultado'] == 'Empate').sum()} ({(df_partidos['resultado'] == 'Empate').mean()*100:.1f}%)")
print(f"- Victorias visitantes: {(df_partidos['resultado'] == 'Visitante').sum()} ({(df_partidos['resultado'] == 'Visitante').mean()*100:.1f}%)")

print(f"\nüìã VISTA PREVIA DEL DATASET:")
print(df_partidos.head())
```

## 4. Pr√°ctica: Gr√°ficas Comparativas (Local vs. Visitante)

Uno de los an√°lisis m√°s importantes en el f√∫tbol es determinar si existe realmente una "ventaja de local". Vamos a crear visualizaciones que nos ayuden a responder esta pregunta.

```python
# 4.1 Gr√°fico de Barras: Comparaci√≥n de Goles Local vs. Visitante
fig, axes = plt.subplots(2, 2, figsize=(15, 12))
fig.suptitle('üèüÔ∏è AN√ÅLISIS DE VENTAJA LOCAL vs. VISITANTE', fontsize=16, fontweight='bold')

# Gr√°fico 1: Goles promedio
ax1 = axes[0, 0]
goles_promedio = [df_partidos['goles_local'].mean(), df_partidos['goles_visitante'].mean()]
colores = ['#1f77b4', '#ff7f0e']
barras = ax1.bar(['Local', 'Visitante'], goles_promedio, color=colores, alpha=0.8)
ax1.set_title('Goles Promedio por Partido', fontweight='bold')
ax1.set_ylabel('Goles Promedio')
ax1.grid(True, alpha=0.3)

# Agregar valores en las barras
for i, v in enumerate(goles_promedio):
    ax1.text(i, v + 0.02, f'{v:.2f}', ha='center', va='bottom', fontweight='bold')

# Gr√°fico 2: Distribuci√≥n de resultados
ax2 = axes[0, 1]
resultados_count = df_partidos['resultado'].value_counts()
colors = ['#2ca02c', '#d62728', '#ff7f0e']
wedges, texts, autotexts = ax2.pie(resultados_count.values, labels=resultados_count.index, 
                                   autopct='%1.1f%%', colors=colors, startangle=90)
ax2.set_title('Distribuci√≥n de Resultados', fontweight='bold')

# Gr√°fico 3: Posesi√≥n promedio
ax3 = axes[1, 0]
posesion_promedio = [df_partidos['posesion_local'].mean(), df_partidos['posesion_visitante'].mean()]
barras = ax3.bar(['Local', 'Visitante'], posesion_promedio, color=colores, alpha=0.8)
ax3.set_title('Posesi√≥n Promedio (%)', fontweight='bold')
ax3.set_ylabel('Posesi√≥n (%)')
ax3.grid(True, alpha=0.3)

# Agregar valores en las barras
for i, v in enumerate(posesion_promedio):
    ax3.text(i, v + 0.5, f'{v:.1f}%', ha='center', va='bottom', fontweight='bold')

# Gr√°fico 4: Tiros a porter√≠a
ax4 = axes[1, 1]
tiros_promedio = [df_partidos['tiros_local'].mean(), df_partidos['tiros_visitante'].mean()]
barras = ax4.bar(['Local', 'Visitante'], tiros_promedio, color=colores, alpha=0.8)
ax4.set_title('Tiros a Porter√≠a Promedio', fontweight='bold')
ax4.set_ylabel('Tiros Promedio')
ax4.grid(True, alpha=0.3)

# Agregar valores en las barras
for i, v in enumerate(tiros_promedio):
    ax4.text(i, v + 0.1, f'{v:.1f}', ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

# An√°lisis estad√≠stico
print("üìä AN√ÅLISIS ESTAD√çSTICO: VENTAJA LOCAL")
print("=" * 50)
print(f"üèüÔ∏è Goles promedio local: {df_partidos['goles_local'].mean():.2f}")
print(f"‚úàÔ∏è Goles promedio visitante: {df_partidos['goles_visitante'].mean():.2f}")
print(f"üìà Diferencia: {(df_partidos['goles_local'].mean() - df_partidos['goles_visitante'].mean()):.2f} goles")

print(f"\nüéØ Posesi√≥n promedio local: {df_partidos['posesion_local'].mean():.1f}%")
print(f"üéØ Posesi√≥n promedio visitante: {df_partidos['posesion_visitante'].mean():.1f}%")

print(f"\n‚öΩ Tiros promedio local: {df_partidos['tiros_local'].mean():.1f}")
print(f"‚öΩ Tiros promedio visitante: {df_partidos['tiros_visitante'].mean():.1f}")

print(f"\nüí° CONCLUSI√ìN: {'S√ç' if df_partidos['goles_local'].mean() > df_partidos['goles_visitante'].mean() else 'NO'} existe ventaja local en los datos")
```

## 5. Visualizaciones Avanzadas con Seaborn

### 5.1 Distribuciones y Patrones

Seaborn nos permite crear visualizaciones m√°s sofisticadas y estad√≠sticamente informativas. Vamos a explorar las distribuciones de goles y buscar patrones interesantes.

```python
# 5.1 Visualizaciones avanzadas con Seaborn
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('üìä AN√ÅLISIS AVANZADO CON SEABORN', fontsize=16, fontweight='bold')

# Gr√°fico 1: Distribuci√≥n de goles por equipo (violin plot)
ax1 = axes[0, 0]
# Preparar datos para violin plot
datos_goles = []
for _, row in df_partidos.iterrows():
    datos_goles.append({'equipo': row['local'], 'goles': row['goles_local'], 'condicion': 'Local'})
    datos_goles.append({'equipo': row['visitante'], 'goles': row['goles_visitante'], 'condicion': 'Visitante'})

df_goles = pd.DataFrame(datos_goles)
sns.violinplot(data=df_goles, x='condicion', y='goles', ax=ax1)
ax1.set_title('Distribuci√≥n de Goles: Local vs. Visitante', fontweight='bold')
ax1.set_ylabel('Goles')

# Gr√°fico 2: Relaci√≥n entre posesi√≥n y goles (scatter plot)
ax2 = axes[0, 1]
# Preparar datos para scatter plot
datos_posesion = []
for _, row in df_partidos.iterrows():
    datos_posesion.append({'posesion': row['posesion_local'], 'goles': row['goles_local']})
    datos_posesion.append({'posesion': row['posesion_visitante'], 'goles': row['goles_visitante']})

df_posesion = pd.DataFrame(datos_posesion)
sns.scatterplot(data=df_posesion, x='posesion', y='goles', ax=ax2, alpha=0.6)
sns.regplot(data=df_posesion, x='posesion', y='goles', ax=ax2, scatter=False, color='red')
ax2.set_title('Relaci√≥n: Posesi√≥n vs. Goles', fontweight='bold')
ax2.set_xlabel('Posesi√≥n (%)')
ax2.set_ylabel('Goles')

# Gr√°fico 3: Heatmap de correlaciones
ax3 = axes[1, 0]
# Preparar datos para correlaci√≥n
datos_correlacion = []
for _, row in df_partidos.iterrows():
    datos_correlacion.append({
        'goles': row['goles_local'],
        'posesion': row['posesion_local'],
        'tiros': row['tiros_local'],
        'faltas': row['faltas_local'],
        'tarjetas': row['tarjetas_local']
    })
    datos_correlacion.append({
        'goles': row['goles_visitante'],
        'posesion': row['posesion_visitante'],
        'tiros': row['tiros_visitante'],
        'faltas': row['faltas_visitante'],
        'tarjetas': row['tarjetas_visitante']
    })

df_correlacion = pd.DataFrame(datos_correlacion)
correlacion_matrix = df_correlacion.corr()
sns.heatmap(correlacion_matrix, annot=True, cmap='coolwarm', center=0, ax=ax3)
ax3.set_title('Matriz de Correlaciones', fontweight='bold')

# Gr√°fico 4: Boxplot de goles por mes
ax4 = axes[1, 1]
df_partidos['mes'] = df_partidos['fecha'].dt.month
df_partidos['mes_nombre'] = df_partidos['fecha'].dt.strftime('%b')
sns.boxplot(data=df_partidos, x='mes_nombre', y='total_goles', ax=ax4)
ax4.set_title('Distribuci√≥n de Goles por Mes', fontweight='bold')
ax4.set_xlabel('Mes')
ax4.set_ylabel('Total de Goles')
ax4.tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# An√°lisis de correlaciones
print("üîç AN√ÅLISIS DE CORRELACIONES")
print("=" * 50)
print("Correlaciones m√°s significativas:")
correlaciones = correlacion_matrix.abs().unstack().sort_values(ascending=False)
correlaciones_filtradas = correlaciones[correlaciones < 1.0]  # Excluir correlaciones perfectas (consigo mismo)
print(correlaciones_filtradas.head(5))

print(f"\nüìà Correlaci√≥n posesi√≥n-goles: {correlacion_matrix.loc['posesion', 'goles']:.3f}")
print(f"üéØ Correlaci√≥n tiros-goles: {correlacion_matrix.loc['tiros', 'goles']:.3f}")
print(f"‚ö†Ô∏è Correlaci√≥n faltas-goles: {correlacion_matrix.loc['faltas', 'goles']:.3f}")
```

## 6. Replicando "La Remontada": An√°lisis Temporal

### 6.1 Simulaci√≥n de un Partido √âpico

Vamos a replicar el tipo de an√°lisis que se har√≠a para un partido legendario como "La Remontada" del PSG vs. Barcelona (6-1, 2017). Crearemos visualizaciones que muestren la evoluci√≥n temporal de un partido.

```python
# 6.1 Simulaci√≥n de "La Remontada" - An√°lisis temporal de un partido √©pico
np.random.seed(95)  # Para crear una "remontada" espec√≠fica

# Simular un partido minuto a minuto
minutos = list(range(1, 91))  # 90 minutos de partido
goles_acumulados_local = []
goles_acumulados_visitante = []
posesion_temporal = []
intensidad_juego = []

# Configurar la "remontada"
goles_local = 0
goles_visitante = 0

# Eventos de gol predefinidos para la narrativa
eventos_gol = {
    12: ('visitante', 'Gol temprano del visitante'),
    25: ('visitante', 'Segundo gol visitante'),
    40: ('visitante', 'Tercer gol visitante - situaci√≥n cr√≠tica'),
    50: ('local', 'Primer gol local - inicia la remontada'),
    65: ('local', 'Segundo gol local - esperanza'),
    75: ('local', 'Tercer gol local - empate √©pico'),
    85: ('local', 'Cuarto gol local - remontada completada'),
    89: ('local', 'Quinto gol local - victoria hist√≥rica')
}

for minuto in minutos:
    # Verificar si hay gol en este minuto
    if minuto in eventos_gol:
        equipo_gol, descripcion = eventos_gol[minuto]
        if equipo_gol == 'local':
            goles_local += 1
        else:
            goles_visitante += 1
    
    # Registrar goles acumulados
    goles_acumulados_local.append(goles_local)
    goles_acumulados_visitante.append(goles_visitante)
    
    # Simular posesi√≥n (cambia seg√∫n el marcador)
    if goles_local > goles_visitante:
        # Equipo local dominando
        posesion_minuto = np.random.normal(60, 10)
    elif goles_visitante > goles_local:
        # Equipo visitante dominando
        posesion_minuto = np.random.normal(40, 10)
    else:
        # Empate
        posesion_minuto = np.random.normal(50, 8)
    
    posesion_minuto = np.clip(posesion_minuto, 20, 80)
    posesion_temporal.append(posesion_minuto)
    
    # Simular intensidad del juego
    intensidad = np.random.normal(50, 20)
    if minuto in eventos_gol:
        intensidad += 30  # Intensidad aumenta con goles
    intensidad = np.clip(intensidad, 0, 100)
    intensidad_juego.append(intensidad)

# Crear DataFrame para el an√°lisis temporal
df_temporal = pd.DataFrame({
    'minuto': minutos,
    'goles_local': goles_acumulados_local,
    'goles_visitante': goles_acumulados_visitante,
    'posesion_local': posesion_temporal,
    'intensidad': intensidad_juego
})

# Crear visualizaci√≥n estilo "La Remontada"
fig, axes = plt.subplots(3, 1, figsize=(16, 14))
fig.suptitle('‚öΩ "LA REMONTADA" - AN√ÅLISIS TEMPORAL DEL PARTIDO', fontsize=18, fontweight='bold')

# Gr√°fico 1: Evoluci√≥n del marcador
ax1 = axes[0]
ax1.plot(df_temporal['minuto'], df_temporal['goles_local'], 
         linewidth=3, marker='o', markersize=8, label='Equipo Local', color='#1f77b4')
ax1.plot(df_temporal['minuto'], df_temporal['goles_visitante'], 
         linewidth=3, marker='s', markersize=8, label='Equipo Visitante', color='#ff7f0e')

# Marcar los momentos de gol
for minuto, (equipo, descripcion) in eventos_gol.items():
    goles_en_minuto = df_temporal[df_temporal['minuto'] == minuto]
    if equipo == 'local':
        ax1.annotate(f"‚öΩ {minuto}'", 
                    xy=(minuto, goles_en_minuto['goles_local'].iloc[0]), 
                    xytext=(minuto, goles_en_minuto['goles_local'].iloc[0] + 0.5),
                    arrowprops=dict(arrowstyle='->', color='blue', lw=2),
                    fontsize=10, fontweight='bold', color='blue')
    else:
        ax1.annotate(f"‚öΩ {minuto}'", 
                    xy=(minuto, goles_en_minuto['goles_visitante'].iloc[0]), 
                    xytext=(minuto, goles_en_minuto['goles_visitante'].iloc[0] + 0.5),
                    arrowprops=dict(arrowstyle='->', color='orange', lw=2),
                    fontsize=10, fontweight='bold', color='orange')

ax1.set_title('Evoluci√≥n del Marcador', fontsize=14, fontweight='bold')
ax1.set_xlabel('Minuto')
ax1.set_ylabel('Goles Acumulados')
ax1.legend()
ax1.grid(True, alpha=0.3)
ax1.set_ylim(-0.5, 6)

# Gr√°fico 2: Posesi√≥n a lo largo del partido
ax2 = axes[1]
ax2.plot(df_temporal['minuto'], df_temporal['posesion_local'], 
         linewidth=2, alpha=0.8, label='Posesi√≥n Local (%)', color='#2ca02c')
ax2.axhline(y=50, color='red', linestyle='--', alpha=0.7, label='Equilibrio (50%)')
ax2.fill_between(df_temporal['minuto'], df_temporal['posesion_local'], 50, 
                 alpha=0.3, color='#2ca02c')
ax2.set_title('Evoluci√≥n de la Posesi√≥n', fontsize=14, fontweight='bold')
ax2.set_xlabel('Minuto')
ax2.set_ylabel('Posesi√≥n Local (%)')
ax2.legend()
ax2.grid(True, alpha=0.3)
ax2.set_ylim(20, 80)

# Gr√°fico 3: Intensidad del juego
ax3 = axes[2]
ax3.plot(df_temporal['minuto'], df_temporal['intensidad'], 
         linewidth=2, alpha=0.8, color='#d62728', label='Intensidad del Juego')
ax3.fill_between(df_temporal['minuto'], df_temporal['intensidad'], 
                 alpha=0.3, color='#d62728')

# Marcar momentos clave
for minuto in eventos_gol.keys():
    ax3.axvline(x=minuto, color='black', linestyle=':', alpha=0.7)

ax3.set_title('Intensidad del Juego', fontsize=14, fontweight='bold')
ax3.set_xlabel('Minuto')
ax3.set_ylabel('Intensidad (0-100)')
ax3.legend()
ax3.grid(True, alpha=0.3)
ax3.set_ylim(0, 100)

plt.tight_layout()
plt.show()

# Estad√≠sticas del partido
print("üìä ESTAD√çSTICAS DEL PARTIDO √âPICO")
print("=" * 50)
print(f"üèÜ Resultado final: {goles_local} - {goles_visitante}")
print(f"üìà Remontada: De 0-3 a {goles_local}-3")
print(f"‚è±Ô∏è Posesi√≥n promedio local: {df_temporal['posesion_local'].mean():.1f}%")
print(f"üî• Intensidad promedio: {df_temporal['intensidad'].mean():.1f}/100")

# An√°lisis por tiempos
primer_tiempo = df_temporal[df_temporal['minuto'] <= 45]
segundo_tiempo = df_temporal[df_temporal['minuto'] > 45]

print(f"\n‚è∞ AN√ÅLISIS POR TIEMPOS:")
print(f"Primer tiempo - Posesi√≥n local: {primer_tiempo['posesion_local'].mean():.1f}%")
print(f"Segundo tiempo - Posesi√≥n local: {segundo_tiempo['posesion_local'].mean():.1f}%")
print(f"Diferencia: {segundo_tiempo['posesion_local'].mean() - primer_tiempo['posesion_local'].mean():.1f}%")

print(f"\nüéØ MOMENTOS CLAVE:")
for minuto, (equipo, descripcion) in eventos_gol.items():
    print(f"‚öΩ Minuto {minuto}: {descripcion}")
```
