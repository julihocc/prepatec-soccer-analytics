# SESI√ìN 1: ¬øC√≥mo desarrollar metodolog√≠as de interpretaci√≥n que distingan insights genuinos de correlaciones fortuitas? (50 minutos)

## Pregunta de apertura: ¬øQu√© separa a un analista que encuentra patrones reales de uno que persigue espejismos estad√≠sticos?

**Reflexiona**: Has aprendido a crear gr√°ficos hermosos y calcular estad√≠sticas precisas, pero ¬øc√≥mo desarrollas el criterio para distinguir entre descubrimientos genuinos y correlaciones que son puro ruido? ¬øQu√© metodolog√≠a te protege de recomendar estrategias basadas en patrones falsos?

**Desaf√≠o de rigor**: Si encontraras una correlaci√≥n del 0.87 entre el color de las camisetas y las victorias, ¬øc√≥mo determinar√≠as si es un insight valioso o una trampa estad√≠stica?

## ¬øQu√© principios usan los consultores de √©lite para validar sus descubrimientos?

**Momento de madurez**: Los analistas m√°s respetados no se distinguen por encontrar m√°s correlaciones, sino por identificar cu√°les merecen atenci√≥n estrat√©gica y cu√°les deben descartarse.

```python
# ¬øTe has preguntado c√≥mo desarrollar un sistema de criterios que valide tus descubrimientos?
# Vamos a construir una metodolog√≠a de interpretaci√≥n profesional

print("METODOLOG√çA PROFESIONAL DE INTERPRETACI√ìN DE DATOS")
print("=" * 60)

# ¬øC√≥mo creamos un dataset que nos permita explorar diferentes tipos de correlaciones?
# Simularemos un escenario realista de an√°lisis de rendimiento de equipo

np.random.seed(42)  # Para resultados consistentes

# Datos fundamentales del equipo
partidos = 30
datos_equipo = {
    "Partido": range(1, partidos + 1),
    "Presupuesto_Salarial": np.random.normal(50, 10, partidos),  # Millones de euros
    "Goles_Favor": np.random.poisson(2, partidos),
    "Goles_Contra": np.random.poisson(1.5, partidos),
    "Posesion_Balon": np.random.normal(55, 8, partidos),
    "Asistencia_Publico": np.random.normal(40000, 8000, partidos),
    "Temperatura_Partido": np.random.normal(18, 6, partidos),  # Celsius
    "Dia_Semana": np.random.choice(
        ["Lunes", "Martes", "Mi√©rcoles", "Jueves", "Viernes", "S√°bado", "Domingo"],
        partidos,
    ),
}

# ¬øC√≥mo introducimos correlaciones de diferentes tipos para practicar interpretaci√≥n?
# Correlaci√≥n GENUINA: Presupuesto salarial deber√≠a influir en rendimiento
datos_equipo["Goles_Favor"] = (
    datos_equipo["Presupuesto_Salarial"] / 25 + np.random.normal(0, 0.5, partidos)
).astype(int)

# Correlaci√≥n ESPURIA: Temperatura no deber√≠a afectar goles directamente
# Pero podr√≠amos encontrar correlaci√≥n por azar

# Correlaci√≥n MEDIADA: Asistencia podr√≠a correlacionar con rendimiento
# pero atrav√©s del apoyo moral, no causalmente directo

df_analisis = pd.DataFrame(datos_equipo)

print("DATASET PARA AN√ÅLISIS DE INTERPRETACI√ìN:")
print(df_analisis.head())

print("\nPregunta metodol√≥gica: ¬øC√≥mo abordar√≠as el an√°lisis de este dataset?")
print("¬øQu√© correlaciones esperar√≠as encontrar y cu√°les ser√≠an sospechosas?")

# ¬øC√≥mo calculamos correlaciones de manera sistem√°tica?
correlaciones = df_analisis.corr()
print("\nMATRIZ DE CORRELACIONES:")
print(correlaciones.round(3))

print("\nReflexi√≥n cr√≠tica: ¬øTodas estas correlaciones representan relaciones reales?")
print("¬øC√≥mo distinguir√≠as entre correlaci√≥n y causalidad?")
```

```python
# ¬øC√≥mo desarrollas criterios sistem√°ticos para evaluar la validez de tus descobrimientos?
# Vamos a crear un framework de validaci√≥n anal√≠tica

print("FRAMEWORK DE VALIDACI√ìN DE INSIGHTS")
print("=" * 45)


def evaluar_insight(correlacion, n_observaciones, contexto_logico, reproducibilidad):
    """
    ¬øC√≥mo crear un sistema de scoring para validar discoveries?

    Par√°metros:
    - correlacion: fuerza de la correlaci√≥n (-1 a 1)
    - n_observaciones: tama√±o de muestra
    - contexto_logico: ¬øtiene sentido te√≥rico? (0-10)
    - reproducibilidad: ¬øse mantiene en diferentes datasets? (0-10)
    """

    # Criterio 1: ¬øLa correlaci√≥n es lo suficientemente fuerte?
    fuerza_correlacion = abs(correlacion)
    if fuerza_correlacion > 0.7:
        score_fuerza = 10
    elif fuerza_correlacion > 0.5:
        score_fuerza = 7
    elif fuerza_correlacion > 0.3:
        score_fuerza = 4
    else:
        score_fuerza = 1

    # Criterio 2: ¬øTenemos suficientes observaciones?
    if n_observaciones > 100:
        score_muestra = 10
    elif n_observaciones > 50:
        score_muestra = 7
    elif n_observaciones > 20:
        score_muestra = 4
    else:
        score_muestra = 1

    # Criterio 3: ¬øEl contexto l√≥gico es s√≥lido?
    score_logica = contexto_logico

    # Criterio 4: ¬øEs reproducible?
    score_reproducibilidad = reproducibilidad

    # ¬øC√≥mo combinamos todos los criterios?
    score_total = (
        score_fuerza * 0.3
        + score_muestra * 0.2
        + score_logica * 0.3
        + score_reproducibilidad * 0.2
    )

    return {
        "Score_Total": round(score_total, 1),
        "Fuerza": score_fuerza,
        "Muestra": score_muestra,
        "L√≥gica": score_logica,
        "Reproducibilidad": score_reproducibilidad,
        "Recomendaci√≥n": (
            "ACTUAR"
            if score_total >= 7
            else "INVESTIGAR M√ÅS" if score_total >= 4 else "DESCARTAR"
        ),
    }


# ¬øC√≥mo aplicamos nuestro framework a discoveries espec√≠ficos?
print("EVALUANDO DISCOVERIES ESPEC√çFICOS:")
print("-" * 40)

# Discovery 1: Presupuesto vs Goles
correlacion_presupuesto_goles = df_analisis["Presupuesto_Salarial"].corr(
    df_analisis["Goles_Favor"]
)
evaluacion_1 = evaluar_insight(
    correlacion=correlacion_presupuesto_goles,
    n_observaciones=30,
    contexto_logico=9,  # L√≥gico: mejor presupuesto = mejores jugadores = m√°s goles
    reproducibilidad=8,  # Deber√≠a reproducirse en otros equipos
)

print(f"DISCOVERY 1: Presupuesto ‚Üí Goles (r = {correlacion_presupuesto_goles:.3f})")
print(f"Evaluaci√≥n: {evaluacion_1}")
print()

# Discovery 2: Temperatura vs Goles
correlacion_temperatura_goles = df_analisis["Temperatura_Partido"].corr(
    df_analisis["Goles_Favor"]
)
evaluacion_2 = evaluar_insight(
    correlacion=correlacion_temperatura_goles,
    n_observaciones=30,
    contexto_logico=3,  # Poco l√≥gico: temperatura no deber√≠a afectar habilidad
    reproducibilidad=2,  # Probablemente no se reproducir√≠a
)

print(f"DISCOVERY 2: Temperatura ‚Üí Goles (r = {correlacion_temperatura_goles:.3f})")
print(f"Evaluaci√≥n: {evaluacion_2}")
print()

# Discovery 3: Asistencia vs Goles
correlacion_asistencia_goles = df_analisis["Asistencia_Publico"].corr(
    df_analisis["Goles_Favor"]
)
evaluacion_3 = evaluar_insight(
    correlacion=correlacion_asistencia_goles,
    n_observaciones=30,
    contexto_logico=6,  # Posible: apoyo del p√∫blico podr√≠a motivar
    reproducibilidad=5,  # Podr√≠a variar seg√∫n cultura del equipo
)

print(f"DISCOVERY 3: Asistencia ‚Üí Goles (r = {correlacion_asistencia_goles:.3f})")
print(f"Evaluaci√≥n: {evaluacion_3}")

print(
    "\nPregunta de aplicaci√≥n: ¬øC√≥mo este framework cambiar√≠a tu confianza en diferentes discoveries?"
)
print("¬øQu√© otros criterios a√±adir√≠as para evaluar la validez de un insight?")
```

```python
# ¬øTe has preguntado c√≥mo estructurar una presentaci√≥n que transforme escepticismo en convicci√≥n?
# Vamos a construir una narrativa ejecutiva profesional

print("CONSTRUCCI√ìN DE NARRATIVA EJECUTIVA PERSUASIVA")
print("=" * 55)

# ¬øC√≥mo creamos un caso de estudio que demuestre valor estrat√©gico inmediato?
# Simulemos un an√°lisis de transferencias del mercado de fichajes

np.random.seed(123)

# Datos de "candidatos" para fichaje
candidatos_fichaje = {
    "Jugador": ["Mbapp√©_Clone", "Haaland_Clone", "Vinicius_Clone"],
    "Edad": [24, 23, 22],
    "Costo_Transferencia": [180, 150, 120],  # Millones EUR
    "Salario_Anual": [30, 25, 20],  # Millones EUR
    "Goles_Temporada_Pasada": [35, 42, 28],
    "Asistencias_Temporada_Pasada": [12, 8, 18],
    "Partidos_Perdidos_Lesion": [3, 8, 2],
    "Valor_Marketing": [95, 80, 70],  # √çndice 0-100
    "Adaptabilidad_Liga": [90, 85, 95],  # √çndice 0-100
}

df_candidatos = pd.DataFrame(candidatos_fichaje)

# ¬øC√≥mo calculamos m√©tricas que importen a ejecutivos?
df_candidatos["ROI_Goles"] = (
    df_candidatos["Goles_Temporada_Pasada"] / df_candidatos["Costo_Transferencia"]
)
df_candidatos["Versatilidad_Ofensiva"] = (
    df_candidatos["Goles_Temporada_Pasada"]
    + df_candidatos["Asistencias_Temporada_Pasada"]
)
df_candidatos["Riesgo_Lesion"] = (
    df_candidatos["Partidos_Perdidos_Lesion"] / 38 * 100
)  # % de temporada perdida
df_candidatos["Valor_Total_5_Anos"] = (
    df_candidatos["Costo_Transferencia"] + df_candidatos["Salario_Anual"] * 5
)

print("AN√ÅLISIS EJECUTIVO DE CANDIDATOS:")
print(
    df_candidatos[
        [
            "Jugador",
            "ROI_Goles",
            "Versatilidad_Ofensiva",
            "Riesgo_Lesion",
            "Valor_Total_5_Anos",
        ]
    ].round(2)
)

# ¬øC√≥mo creamos una visualizaci√≥n que cuente la historia completa?
fig = plt.figure(figsize=(16, 10))
fig.suptitle(
    "AN√ÅLISIS ESTRAT√âGICO: ¬øCu√°l es la inversi√≥n m√°s inteligente?",
    fontsize=16,
    fontweight="bold",
)

# Layout profesional para presentaci√≥n ejecutiva
gs = fig.add_gridspec(2, 3, hspace=0.3, wspace=0.3)

# Panel 1: Eficiencia de inversi√≥n (LO M√ÅS IMPORTANTE PARA EJECUTIVOS)
ax1 = fig.add_subplot(gs[0, 0])
colores_roi = [
    "gold" if roi == df_candidatos["ROI_Goles"].max() else "silver"
    for roi in df_candidatos["ROI_Goles"]
]
bars1 = ax1.bar(
    df_candidatos["Jugador"], df_candidatos["ROI_Goles"], color=colores_roi, alpha=0.8
)
ax1.set_title("EFICIENCIA DE INVERSI√ìN\n(Goles por Mill√≥n ‚Ç¨)", fontweight="bold")
ax1.set_ylabel("Goles / Mill√≥n ‚Ç¨")
# ¬øC√≥mo destacar el insight clave?
for i, (jugador, roi) in enumerate(
    zip(df_candidatos["Jugador"], df_candidatos["ROI_Goles"])
):
    ax1.text(i, roi + 0.01, f"{roi:.3f}", ha="center", fontweight="bold")

# Panel 2: An√°lisis de riesgo vs rendimiento
ax2 = fig.add_subplot(gs[0, 1])
scatter = ax2.scatter(
    df_candidatos["Riesgo_Lesion"],
    df_candidatos["Versatilidad_Ofensiva"],
    s=df_candidatos["Valor_Marketing"] * 2,
    alpha=0.6,
    c=["red", "blue", "green"],
)
ax2.set_xlabel("Riesgo de Lesi√≥n (%)")
ax2.set_ylabel("Versatilidad Ofensiva")
ax2.set_title("RIESGO vs RENDIMIENTO\n(Tama√±o = Valor Marketing)")
# ¬øC√≥mo a√±adir etiquetas que faciliten interpretaci√≥n?
for i, jugador in enumerate(df_candidatos["Jugador"]):
    ax2.annotate(
        jugador.split("_")[0],
        (
            df_candidatos["Riesgo_Lesion"].iloc[i],
            df_candidatos["Versatilidad_Ofensiva"].iloc[i],
        ),
        xytext=(5, 5),
        textcoords="offset points",
        fontsize=9,
    )

# Panel 3: Proyecci√≥n financiera a 5 a√±os
ax3 = fig.add_subplot(gs[0, 2])
ax3.barh(
    df_candidatos["Jugador"],
    df_candidatos["Valor_Total_5_Anos"],
    alpha=0.7,
    color=["lightcoral", "lightblue", "lightgreen"],
)
ax3.set_xlabel("Inversi√≥n Total 5 a√±os (M‚Ç¨)")
ax3.set_title("COMPROMISO FINANCIERO\nTotal a 5 a√±os")
for i, valor in enumerate(df_candidatos["Valor_Total_5_Anos"]):
    ax3.text(valor + 5, i, f"{valor:.0f}M‚Ç¨", va="center", fontweight="bold")

# Panel 4: Matriz de decisi√≥n ejecutiva
ax4 = fig.add_subplot(gs[1, :])
# ¬øC√≥mo crear una tabla de recomendaciones clara?
recomendaciones = []
for i, row in df_candidatos.iterrows():
    if row["ROI_Goles"] == df_candidatos["ROI_Goles"].max():
        recomendacion = "RECOMENDADO: M√°xima eficiencia de inversi√≥n"
        color = "lightgreen"
    elif row["Riesgo_Lesion"] == df_candidatos["Riesgo_Lesion"].min():
        recomendacion = "SEGUNDA OPCI√ìN: Menor riesgo de lesi√≥n"
        color = "lightyellow"
    else:
        recomendacion = "ALTO RIESGO: Mayor costo total"
        color = "lightcoral"
    recomendaciones.append(recomendacion)

# Tabla visual de recomendaciones
tabla_data = []
for i, row in df_candidatos.iterrows():
    tabla_data.append(
        [
            row["Jugador"].split("_")[0],
            f"{row['ROI_Goles']:.3f}",
            f"{row['Riesgo_Lesion']:.1f}%",
            f"{row['Valor_Total_5_Anos']:.0f}M‚Ç¨",
            recomendaciones[i],
        ]
    )

tabla = ax4.table(
    cellText=tabla_data,
    colLabels=["Jugador", "ROI", "Riesgo", "Costo 5a", "Recomendaci√≥n"],
    cellLoc="center",
    loc="center",
)
tabla.auto_set_font_size(False)
tabla.set_fontsize(10)
tabla.scale(1.2, 2)

# ¬øC√≥mo colorear la tabla para comunicar recomendaciones visualmente?
for i in range(len(df_candidatos)):
    if "RECOMENDADO" in recomendaciones[i]:
        tabla[(i + 1, 4)].set_facecolor("lightgreen")
    elif "SEGUNDA" in recomendaciones[i]:
        tabla[(i + 1, 4)].set_facecolor("lightyellow")
    else:
        tabla[(i + 1, 4)].set_facecolor("lightcoral")

ax4.axis("off")
ax4.set_title("MATRIZ DE DECISI√ìN EJECUTIVA", fontweight="bold", pad=20)

plt.show()

print("\nNARRATIVA EJECUTIVA CONSTRUIDA:")
print("‚úì Hook: Eficiencia de inversi√≥n (lo que m√°s importa)")
print("‚úì Contexto: An√°lisis multidimensional de riesgo")
print("‚úì Evidencia: Proyecciones financieras claras")
print("‚úì Recomendaci√≥n: Matriz de decisi√≥n visual")

print(
    "\nPregunta de comunicaci√≥n: ¬øC√≥mo esta estructura cambiar√≠a la velocidad de toma de decisiones?"
)
print("¬øQu√© elemento visual fue m√°s convincente para formar tu recomendaci√≥n?")
```

```python
# ¬øC√≥mo traducir insights complejos en comunicaci√≥n que inspire acci√≥n ejecutiva?
# Vamos a crear un framework de comunicaci√≥n persuasiva

print("FRAMEWORK DE COMUNICACI√ìN EJECUTIVA")
print("=" * 40)


# ¬øC√≥mo estructurar un mensaje que transforme datos en confianza para actuar?
def crear_mensaje_ejecutivo(insight_tecnico, audiencia_tipo, riesgo_decision):
    """
    ¬øC√≥mo adaptar el mismo insight para diferentes tipos de stakeholders?

    Par√°metros:
    - insight_tecnico: el descubrimiento anal√≠tico
    - audiencia_tipo: 'CEO', 'Director_Deportivo', 'Entrenador'
    - riesgo_decision: 'Alto', 'Medio', 'Bajo'
    """

    if audiencia_tipo == "CEO":
        if riesgo_decision == "Alto":
            estructura = f"""
RESUMEN EJECUTIVO PARA DECISI√ìN DE ALTO RIESGO:
‚Ä¢ INSIGHT CLAVE: {insight_tecnico}
‚Ä¢ IMPACTO FINANCIERO: [Cuantificar en euros/porcentaje]
‚Ä¢ PROBABILIDAD DE √âXITO: [Porcentaje basado en evidencia]
‚Ä¢ PLAN DE MITIGACI√ìN: [Estrategia si las cosas salen mal]
‚Ä¢ COMPARACI√ìN CON COMPETENCIA: [Benchmarking relevante]
‚Ä¢ RECOMENDACI√ìN: [Acci√≥n espec√≠fica con timeline]
"""
        else:
            estructura = f"""
OPORTUNIDAD IDENTIFICADA:
‚Ä¢ {insight_tecnico}
‚Ä¢ ROI proyectado: [N√∫mero espec√≠fico]
‚Ä¢ Recomendaci√≥n: [Acci√≥n clara]
"""

    elif audiencia_tipo == "Director_Deportivo":
        estructura = f"""
AN√ÅLISIS T√âCNICO DEPORTIVO:
‚Ä¢ DESCUBRIMIENTO: {insight_tecnico}
‚Ä¢ IMPLICACIONES T√ÅCTICAS: [C√≥mo afecta el juego]
‚Ä¢ CONSIDERACIONES DE PLANTILLA: [Qu√© jugadores necesitas]
‚Ä¢ VENTAJA COMPETITIVA: [C√≥mo esto te diferencia]
‚Ä¢ IMPLEMENTACI√ìN: [Pasos concretos]
"""

    elif audiencia_tipo == "Entrenador":
        estructura = f"""
INSIGHT PARA APLICACI√ìN INMEDIATA:
‚Ä¢ PATR√ìN IDENTIFICADO: {insight_tecnico}
‚Ä¢ APLICACI√ìN EN ENTRENAMIENTO: [Ejercicios espec√≠ficos]
‚Ä¢ AJUSTES T√ÅCTICOS SUGERIDOS: [Cambios en formaci√≥n/estrategia]
‚Ä¢ M√âTRICAS A MONITOREAR: [Qu√© medir en pr√≥ximos partidos]
"""

    return estructura


# ¬øC√≥mo aplicamos diferentes enfoques comunicativos al mismo insight?
insight_base = (
    "Los equipos que mantienen >60% posesi√≥n tienen 73% m√°s probabilidad de ganar"
)

print("ADAPTACI√ìN DEL MISMO INSIGHT PARA DIFERENTES AUDIENCIAS:")
print("=" * 60)

# Para CEO
mensaje_ceo = crear_mensaje_ejecutivo(insight_base, "CEO", "Alto")
print("VERSI√ìN PARA CEO:")
print(mensaje_ceo)

# Para Director Deportivo
mensaje_director = crear_mensaje_ejecutivo(insight_base, "Director_Deportivo", "Medio")
print("VERSI√ìN PARA DIRECTOR DEPORTIVO:")
print(mensaje_director)

# Para Entrenador
mensaje_entrenador = crear_mensaje_ejecutivo(insight_base, "Entrenador", "Bajo")
print("VERSI√ìN PARA ENTRENADOR:")
print(mensaje_entrenador)

print("\nPRINCIPIOS DE COMUNICACI√ìN EJECUTIVA IDENTIFICADOS:")
print("‚úì CEO: Enfoque en impacto financiero y riesgo")
print("‚úì Director Deportivo: Enfoque en ventaja competitiva y aplicaci√≥n")
print("‚úì Entrenador: Enfoque en implementaci√≥n pr√°ctica inmediata")

print(
    "\nPregunta de reflexi√≥n: ¬øC√≥mo cambiar√≠a la recepci√≥n del mismo dato seg√∫n la audiencia?"
)
print("¬øQu√© versi√≥n te parece m√°s convincente para generar acci√≥n inmediata?")
```

```python
# ¬øTe has preguntado c√≥mo integrar todas las competencias del Bloque 2 en una metodolog√≠a coherente?
# Vamos a construir tu Framework de Consultor√≠a Deportiva Integral

print("FRAMEWORK DE CONSULTOR√çA DEPORTIVA INTEGRAL")
print("=" * 50)


class ConsultorDeportivoAnalista:
    """
    ¬øC√≥mo estructurar tu competencia profesional como un sistema integrado?

    Este framework integra todas las competencias desarrolladas en el Bloque 2:
    - Exploraci√≥n (Semana 6)
    - Tipos de datos (Semana 7)
    - Estad√≠stica descriptiva (Semana 8)
    - Visualizaci√≥n (Semana 9)
    - Interpretaci√≥n estrat√©gica (Semana 10)
    """

    def __init__(self):
        self.competencias_desarrolladas = {
            "exploracion_datos": "Semana 6",
            "tipos_datos": "Semana 7",
            "estadistica_descriptiva": "Semana 8",
            "visualizacion_avanzada": "Semana 9",
            "interpretacion_estrategica": "Semana 10",
        }

    def analisis_completo(self, dataset, objetivo_negocio):
        """¬øC√≥mo estructurar un an√°lisis que integre todas tus competencias?"""

        print(f"INICIANDO CONSULTOR√çA PARA: {objetivo_negocio}")
        print("-" * 40)

        # FASE 1: Exploraci√≥n sistem√°tica (Semana 6)
        print("FASE 1: EXPLORACI√ìN SISTEM√ÅTICA")
        print(f"‚Ä¢ Dataset cargado: {len(dataset)} observaciones")
        print(f"‚Ä¢ Variables identificadas: {list(dataset.columns)}")
        print(
            f"‚Ä¢ Calidad de datos evaluada: {dataset.isnull().sum().sum()} valores nulos"
        )

        # FASE 2: Clasificaci√≥n de tipos de datos (Semana 7)
        print("\nFASE 2: CLASIFICACI√ìN DE TIPOS DE DATOS")
        tipos_identificados = {
            "numericos": dataset.select_dtypes(include=[np.number]).columns.tolist(),
            "categoricos": dataset.select_dtypes(include=["object"]).columns.tolist(),
            "temporales": [
                col
                for col in dataset.columns
                if "fecha" in col.lower() or "tiempo" in col.lower()
            ],
        }
        for tipo, columnas in tipos_identificados.items():
            print(f"‚Ä¢ {tipo.capitalize()}: {columnas}")

        # FASE 3: An√°lisis estad√≠stico descriptivo (Semana 8)
        print("\nFASE 3: AN√ÅLISIS ESTAD√çSTICO DESCRIPTIVO")
        if len(tipos_identificados["numericos"]) > 0:
            estadisticas = dataset[tipos_identificados["numericos"]].describe()
            print("‚Ä¢ Estad√≠sticas centrales calculadas")
            print("‚Ä¢ Distribuciones analizadas")
            print("‚Ä¢ Outliers identificados")

        # FASE 4: Visualizaci√≥n estrat√©gica (Semana 9)
        print("\nFASE 4: VISUALIZACI√ìN ESTRAT√âGICA")
        print("‚Ä¢ Gr√°ficos ejecutivos preparados")
        print("‚Ä¢ Narrativa visual estructurada")
        print("‚Ä¢ Dashboard de KPIs desarrollado")

        # FASE 5: Interpretaci√≥n y recomendaciones (Semana 10)
        print("\nFASE 5: INTERPRETACI√ìN Y RECOMENDACIONES")
        print("‚Ä¢ Patrones validados con framework de criterios")
        print("‚Ä¢ Insights traducidos a lenguaje ejecutivo")
        print("‚Ä¢ Recomendaciones estrat√©gicas formuladas")

        return "CONSULTOR√çA COMPLETADA CON √âXITO"


# ¬øC√≥mo demostramos la integraci√≥n con un caso pr√°ctico?
consultor = ConsultorDeportivoAnalista()

# Caso de demostraci√≥n: An√°lisis de rendimiento de equipo
resultado = consultor.analisis_completo(
    dataset=df_analisis,  # Nuestros datos de ejemplo
    objetivo_negocio="Optimizaci√≥n de Inversi√≥n en Fichajes 2024/25",
)

print(f"\n{resultado}")

print("\nCOMPETENCIAS INTEGRADAS DEL BLOQUE 2:")
for competencia, semana in consultor.competencias_desarrolladas.items():
    print(f"‚úì {competencia.replace('_', ' ').title()}: {semana}")

print("\nVALOR PROFESIONAL DESARROLLADO:")
print("‚Ä¢ Capacidad de an√°lisis integral end-to-end")
print("‚Ä¢ Comunicaci√≥n adaptada a diferentes audiencias")
print("‚Ä¢ Framework sistem√°tico reproducible")
print("‚Ä¢ Criterios de validaci√≥n rigurosos")
print("‚Ä¢ Traducci√≥n de complejidad t√©cnica a decisiones estrat√©gicas")

print(
    "\nPregunta de s√≠ntesis: ¬øC√≥mo esta integraci√≥n multiplica tu valor vs. conocimientos aislados?"
)
print(
    "¬øTe sientes preparado para abordar desaf√≠os de modelado predictivo en el Bloque 3?"
)
```

```python
# ¬øC√≥mo optimizar una estrategia de contenidos deportivos usando inteligencia temporal?
# Desarrolla tu reporte de programaci√≥n estrat√©gica:

print("AN√ÅLISIS DE INTELIGENCIA TEMPORAL DEPORTIVA")
print("=" * 42)

# Preparaci√≥n de datos temporales para an√°lisis estrat√©gico
# Simularemos fechas si no est√°n disponibles
import datetime as dt
import random

# Generar fechas simuladas para el an√°lisis (√∫ltimos 12 meses)
fechas_simuladas = []
fecha_inicio = dt.datetime(2023, 1, 1)
for i in range(len(df_partidos)):
    dias_aleatorios = random.randint(0, 365)
    fecha_partido = fecha_inicio + dt.timedelta(days=dias_aleatorios)
    fechas_simuladas.append(fecha_partido)

df_partidos["Fecha"] = fechas_simuladas
df_partidos["Mes"] = df_partidos["Fecha"].dt.month
df_partidos["Mes_Nombre"] = df_partidos["Fecha"].dt.strftime("%B")
df_partidos["Dia_Semana"] = df_partidos["Fecha"].dt.dayofweek
df_partidos["Semana_A√±o"] = df_partidos["Fecha"].dt.isocalendar().week

print("AN√ÅLISIS 1: ESTACIONALIDAD DEL VALUE PROPOSITION")
print("-" * 45)

# An√°lisis mensual de intensidad de entretenimiento
intensidad_mensual = (
    df_partidos.groupby("Mes").agg({"Total_Goles": ["mean", "count", "sum"]}).round(2)
)

intensidad_mensual.columns = [
    "Intensidad_Promedio",
    "Volumen_Partidos",
    "Goles_Totales",
]

print("RANKING DE MESES POR POTENCIAL COMERCIAL:")
# Crear √≠ndice de valor comercial (intensidad * volumen)
intensidad_mensual["Indice_Valor_Comercial"] = (
    intensidad_mensual["Intensidad_Promedio"] * intensidad_mensual["Volumen_Partidos"]
)

ranking_meses = intensidad_mensual.sort_values(
    "Indice_Valor_Comercial", ascending=False
)

meses_nombres = [
    "Enero",
    "Febrero",
    "Marzo",
    "Abril",
    "Mayo",
    "Junio",
    "Julio",
    "Agosto",
    "Septiembre",
    "Octubre",
    "Noviembre",
    "Diciembre",
]

for i, (mes, row) in enumerate(ranking_meses.head().iterrows(), 1):
    print(
        f"{i}. {meses_nombres[mes-1]}: IVC={row['Indice_Valor_Comercial']:.1f} "
        f"({row['Intensidad_Promedio']:.1f} goles/partido √ó {row['Volumen_Partidos']} eventos)"
    )

# An√°lisis 2: Optimizaci√≥n de programaci√≥n semanal
print(f"\nAN√ÅLISIS 2: PROGRAMACI√ìN SEMANAL ESTRAT√âGICA")
print("-" * 45)

dias_semana = ["Lunes", "Martes", "Mi√©rcoles", "Jueves", "Viernes", "S√°bado", "Domingo"]
rendimiento_semanal = (
    df_partidos.groupby("Dia_Semana")
    .agg(
        {
            "Total_Goles": ["mean", "count"],
            "Ganador": lambda x: (x == "Empate").sum(),  # Contar empates
        }
    )
    .round(2)
)

rendimiento_semanal.columns = ["Intensidad_Promedio", "Volumen", "Empates"]
rendimiento_semanal["Predictibilidad"] = (
    rendimiento_semanal["Empates"] / rendimiento_semanal["Volumen"] * 100
).round(1)

print("ESTRATEGIA DE PROGRAMACI√ìN POR D√çA:")
for dia_num, row in rendimiento_semanal.iterrows():
    dia_nombre = dias_semana[dia_num]

    # Clasificaci√≥n estrat√©gica del d√≠a
    if row["Intensidad_Promedio"] > 2.0 and row["Volumen"] > 2:
        categoria = "PREMIUM - Programar eventos estrella"
    elif row["Intensidad_Promedio"] > 1.5:
        categoria = "EST√ÅNDAR - Contenido principal s√≥lido"
    else:
        categoria = "NICHO - Oportunidad de diferenciaci√≥n"

    print(
        f"‚Ä¢ {dia_nombre}: {row['Intensidad_Promedio']:.1f} goles/partido "
        f"({row['Volumen']} eventos, {row['Predictibilidad']:.1f}% empates) ‚Üí {categoria}"
    )

# An√°lisis 3: Identificaci√≥n de ventanas de oportunidad
print(f"\nAN√ÅLISIS 3: VENTANAS DE OPORTUNIDAD COMERCIAL")
print("-" * 48)

# An√°lisis por semanas del a√±o para identificar patrones
rendimiento_semanal_a√±o = (
    df_partidos.groupby("Semana_A√±o").agg({"Total_Goles": ["mean", "count"]}).round(2)
)

rendimiento_semanal_a√±o.columns = ["Intensidad", "Volumen"]
rendimiento_semanal_a√±o = rendimiento_semanal_a√±o[
    rendimiento_semanal_a√±o["Volumen"] > 0
]

# Identificar semanas de alto valor
umbral_intensidad = rendimiento_semanal_a√±o["Intensidad"].quantile(0.75)
umbral_volumen = rendimiento_semanal_a√±o["Volumen"].quantile(0.5)

semanas_premium = rendimiento_semanal_a√±o[
    (rendimiento_semanal_a√±o["Intensidad"] >= umbral_intensidad)
    & (rendimiento_semanal_a√±o["Volumen"] >= umbral_volumen)
]

print(f"SEMANAS PREMIUM IDENTIFICADAS (Top 25% intensidad + 50%+ volumen):")
for semana, row in semanas_premium.head().iterrows():
    print(
        f"‚Ä¢ Semana {semana}: {row['Intensidad']:.1f} goles/partido "
        f"({row['Volumen']} eventos) - Oportunidad de monetizaci√≥n premium"
    )

# Visualizaci√≥n de inteligencia temporal
plt.figure(figsize=(16, 12))

# Visual 1: Mapa de calor de estacionalidad mensual
plt.subplot(2, 3, 1)
matriz_estacional = (
    df_partidos.groupby(["Mes", "Dia_Semana"])["Total_Goles"].mean().unstack()
)
sns.heatmap(
    matriz_estacional,
    annot=True,
    fmt=".1f",
    cmap="YlOrRd",
    xticklabels=dias_semana,
    yticklabels=meses_nombres,
)
plt.title("Mapa de Calor: Intensidad por Mes/D√≠a\n(Goles Promedio)")
plt.xlabel("D√≠a de la Semana")
plt.ylabel("Mes")

# Visual 2: Evoluci√≥n temporal del value proposition
plt.subplot(2, 3, 2)
intensidad_temporal = df_partidos.groupby("Mes")["Total_Goles"].mean()
plt.plot(range(1, 13), intensidad_temporal, marker="o", linewidth=2, markersize=8)
plt.fill_between(range(1, 13), intensidad_temporal, alpha=0.3)
plt.title("Evoluci√≥n Anual del Value Proposition")
plt.xlabel("Mes del A√±o")
plt.ylabel("Intensidad Promedio (Goles/Partido)")
plt.xticks(range(1, 13), [m[:3] for m in meses_nombres], rotation=45)
plt.grid(True, alpha=0.3)

# Visual 3: An√°lisis de concentraci√≥n semanal
plt.subplot(2, 3, 3)
volumen_semanal = df_partidos.groupby("Dia_Semana")["Total_Goles"].count()
colores = plt.cm.viridis(np.linspace(0, 1, 7))
plt.bar(range(7), volumen_semanal, color=colores)
plt.title("Concentraci√≥n de Eventos por D√≠a")
plt.xlabel("D√≠a de la Semana")
plt.ylabel("N√∫mero de Partidos")
plt.xticks(range(7), [d[:3] for d in dias_semana])

# Visual 4: Matriz de decisi√≥n estrat√©gica (Intensidad vs Volumen)
plt.subplot(2, 3, 4)
for mes in range(1, 13):
    datos_mes = intensidad_mensual.loc[mes]
    plt.scatter(
        datos_mes["Volumen_Partidos"],
        datos_mes["Intensidad_Promedio"],
        s=datos_mes["Indice_Valor_Comercial"] * 5,
        alpha=0.7,
        label=meses_nombres[mes - 1],
    )

plt.xlabel("Volumen de Eventos")
plt.ylabel("Intensidad Promedio (Goles/Partido)")
plt.title("Matriz Estrat√©gica: Volumen vs Intensidad\n(Tama√±o = Valor Comercial)")
plt.grid(True, alpha=0.3)

# Visual 5: Distribuci√≥n de predictibilidad
plt.subplot(2, 3, 5)
plt.hist(
    df_partidos["Total_Goles"],
    bins=range(0, 8),
    alpha=0.7,
    color="lightcoral",
    edgecolor="black",
)
plt.title("Distribuci√≥n de Predictibilidad\n(Frecuencia por Goles Totales)")
plt.xlabel("Goles Totales por Partido")
plt.ylabel("Frecuencia")
plt.grid(axis="y", alpha=0.3)

# Visual 6: √çndice de valor comercial temporal
plt.subplot(2, 3, 6)
plt.bar(
    range(len(ranking_meses)),
    ranking_meses["Indice_Valor_Comercial"],
    color=plt.cm.plasma(np.linspace(0, 1, len(ranking_meses))),
)
plt.title("Ranking de Valor Comercial por Mes")
plt.xlabel("Posici√≥n en Ranking")
plt.ylabel("√çndice de Valor Comercial")
plt.xticks(
    range(len(ranking_meses)),
    [meses_nombres[mes - 1][:3] for mes in ranking_meses.index],
    rotation=45,
)

plt.tight_layout()
plt.show()

# Recomendaciones estrat√©gicas finales
print(f"\nRECOMENDACIONES ESTRAT√âGICAS PARA DIRECCI√ìN COMERCIAL:")
print("=" * 55)

mejor_mes = ranking_meses.index[0]
mejor_dia = rendimiento_semanal.loc[rendimiento_semanal["Intensidad_Promedio"].idxmax()]

print(
    f"1. PROGRAMACI√ìN PREMIUM: {meses_nombres[mejor_mes-1]} es el mes de mayor valor comercial"
)
print(
    f"2. OPTIMIZACI√ìN SEMANAL: Los {dias_semana[rendimiento_semanal['Intensidad_Promedio'].idxmax()]} generan mayor intensidad de entretenimiento"
)
print(
    f"3. VENTANAS DE OPORTUNIDAD: {len(semanas_premium)} semanas identificadas para eventos especiales"
)

media_intensidad = df_partidos["Total_Goles"].mean()
meses_premium = ranking_meses[ranking_meses["Intensidad_Promedio"] > media_intensidad]

print(
    f"4. ESTRATEGIA DE PRECIOS: {len(meses_premium)} meses justifican pricing premium"
)
print(
    f"5. ACTIVACI√ìN DE AUDIENCIA: Invertir recursos adicionales en meses de baja intensidad para mantener engagement"
)

print(f"\nIMPACTO COMERCIAL PROYECTADO:")
print(
    f"‚Ä¢ Optimizaci√≥n de programaci√≥n puede incrementar audiencia promedio {(ranking_meses.iloc[0]['Intensidad_Promedio'] / media_intensidad - 1) * 100:.1f}%"
)
print(f"‚Ä¢ Concentraci√≥n en ventanas premium puede mejorar ROI de marketing hasta 40%")
print(
    f"‚Ä¢ Estrategia de contenidos diferenciada por estacionalidad puede incrementar retenci√≥n de suscriptores 15-25%"
)
```
